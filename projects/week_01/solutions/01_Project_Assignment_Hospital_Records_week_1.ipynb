## Phase 2 - Clean the Messy Fields `[16 marks]`

You now have a list of raw record strings. In this phase you will

### Task 2.1 - Standardise Patient IDs `[3 marks]`

The IDs arrive as `P-001`, `P 002`, `P003` - inconsistent separators.
Standardise all of them to a simple `PXXX` format: the letter P followed
by exactly the digits, `**no spaces or hyphens**`.

Build `clean_ids`: a list of 20 standardised ID strings.

Print all 20 in one line to verify.

**Expected:** `['P001', 'P002', 'P003', 'P004', ...]`

*Hint: `re.sub(r'[^P0-9]', '', pid)` removes everything that is
not `P` or a digit from the ID string.*


````python
# Task 2.1 - Standardise patient IDs
clean_ids = []

for record in records:
    fields = record.split(',')
    pid    = fields[0]
    # --- your cleaning code here ---
    clean_pid = re.sub(r'[^P0-9]', '', pid) #remember?!! re.sub(pattern, replacement, text), finds pattern in text and replaces it with replacement.
    clean_ids.append(clean_pid)

print(clean_ids)
````

### Task 2.2 - Clean Patient Names `[3 marks]`

Names have leading/trailing whitespace, internal double-spaces, and
inconsistent capitalisation (`AHMAD RAZA`, `sara khan`, `ALI hassan`).

Apply this three-step chain to every name:
1. `.strip()` - remove leading/trailing whitespace
2. `' '.join(name.split())` - collapse internal spaces
3. `.title()` - proper title case

Build `clean_names`: a list of 20 cleaned name strings.

Print all 20.

**Expected:** `['Ahmad Raza', 'Sara Khan', 'Bilal Ahmed', ...]`


````python
# Task 2.2 - Clean patient names
clean_names = []

for record in records:
    fields = record.split(',')
    name   = fields[1]
    # --- your cleaning code here ---
    clean_name = ' '.join(name.strip().split()).title()     #remove leading/trailing spaces and split on sapces then title case and apply join with a white space
    clean_names.append(clean_name)

print(clean_names)
````

### Task 2.3 - Extract Numeric Ages `[4 marks]`

Ages arrive in four different formats:

| Raw value | What you want |
|---|---|
| `"34"` | `34` |
| `"28 years"` | `28` |
| `"51yo"` | `51` |
| `"N/A"` | `None` |

Use `re.search(r'\d+', age_str)` to find the numeric part.
If no digits are found, store `None`.

Build `clean_ages`: a list of 20 integers (or `None`).

Then print:
- All 20 values
- How many records have a missing age (`None`)
- The ages of the youngest and oldest patient (ignore `None`)

*Hint: `re.search(r'\d+', s)` returns a match object or `None`.
If a match is found, `match.group()` gives you the matched string,
which you then convert to `int()`.*


````python
# Task 2.3 - Extract numeric ages
clean_ages = []

for record in records:
    fields  = record.split(',')
    age_str = fields[2].strip()
    # --- your cleaning code here ---
    m = re.search(r'\d+', age_str)
    clean_age = int(m.group()) if m else None   # int or None
    clean_ages.append(clean_age)

print("All ages:", clean_ages)
print("Missing ages:", clean_ages.count(None))

valid_ages_temp = [a for a in clean_ages if a is not None]
print(f"Youngest patient: {min(valid_ages_temp)}")
print(f"Oldest patient  : {max(valid_ages_temp)}")
````

### Task 2.4 - Standardise Gender `[3 marks]`

Six different representations all mean either Male or Female:

| Raw values that mean Male | Raw values that mean Female |
|---|---|
| `Male`, `male`, `M`, `m`, `MALE` | `Female`, `female`, `F`, `f`, `FEMALE` |

Rule: if the value (lowercased) starts with `'m'` → `'M'`, otherwise → `'F'`.

Build `clean_genders`: a list of 20 strings, each either `'M'` or `'F'`.

Print all 20 and the count of males and females.


````python
# Task 2.4 - Standardise gender
clean_genders = []

for record in records:
    fields = record.split(',')
    g      = fields[3].strip()
    # --- your cleaning code here ---
    clean_g = 'M' if g.lower().startswith('m') else 'F'    
    clean_genders.append(clean_g)

# male/female counts
print("Genders:", clean_genders)
print(f"Males  : {clean_genders.count('M')}")
print(f"Females: {clean_genders.count('F')}")
````

### Task 2.5 - Convert Temperatures to Celsius `[3 marks]`

Temperatures are recorded in two units:
- `"38.9C"` - already in Celsius, just extract the float
- `"99.0F"` - Fahrenheit, must convert: `°C = (°F − 32) × 5/9`

Steps for each temperature string:
1. Check the last character: `temp[-1]`
2. Extract the numeric value: `float(temp[:-1])`
3. If Fahrenheit (`'F'`), convert; if Celsius (`'C'`), keep as is
4. Round to 1 decimal place

Build `clean_temps`: a list of 20 floats (all in Celsius).

Print all 20 and identify the highest and lowest temperatures.

*Hint: `'99.0F'[:-1]` gives `'99.0'` and `'99.0F'[-1]` gives `'F'`.*


````python
# Task 2.5 - Convert temperatures to Celsius
clean_temps = []

for record in records:
    fields = record.split(',')
    temp   = fields[7].strip()
    # --- your cleaning code here ---
    unit   = temp[-1]
    value  = float(temp[:-1])
    if unit == 'F':
        celsius = (value - 32) * 5 / 9   # conversion formula here
    else:
        celsius = value
    clean_temps.append(round(celsius, 1))

print("Temperatures (C):", clean_temps)
print(f"Highest temperature: {max(clean_temps)}C")
print(f"Lowest temperature : {min(clean_temps)}C")
````

## Phase 3 - Validate with Regex `[12 marks]`

Cleaning fixed the formatting. Validation checks whether the *content* is

### Task 3.1 - Validate Phone Numbers `[4 marks]`

A valid Pakistani mobile number follows one of these two formats:
- `0XXX-XXXXXXX` - 4-digit prefix, hyphen, 7-digit number
- `0XXX XXXXXXX` - same but space separator

Pattern: starts with `0`, then 3 digits, then one separator (`-` or space),
then exactly 7 digits, and that's it - nothing else.

Build two lists:
- `valid_phones` - list of valid phone strings
- `invalid_records` - list of **patient IDs** whose phone is invalid

Print both lists and the count of invalid records.

*Hint: `re.match(r'^0\d{3}[-\s]\d{7}$', phone)` - the `^` and `$` anchor
the match to the whole string so partial matches don't sneak through.*


````python
# Task 3.1 - Validate phone numbers
phone_pattern = re.compile(r'^0\d{3}[-\s]\d{7}$')

valid_phones   = []
invalid_records = []

for i, record in enumerate(records):
    fields = record.split(',')
    phone  = fields[4].strip()
    pid    = clean_ids[i]   # use the clean ID from Phase 2
    # --- your validation code here ---
    if phone_pattern.match(phone):
        valid_phones.append(phone)
    else:
        invalid_records.append(pid)

print("Valid phones:", valid_phones)
print("\nInvalid patient IDs:", invalid_records)
print(f"Invalid count: {len(invalid_records)} out of {len(records)}")
````

### Task 3.2 - Validate Patient ID Format `[4 marks]`

After cleaning, all IDs should be in `PXXX` format:
- Starts with the letter `P`
- Followed by exactly 3 digits
- Nothing else

Check `clean_ids` (from Task 2.1) against this pattern and identify any IDs
that still do not match `PXXX`.

Also answer: was your Task 2.1 cleaning robust enough? Are all 20 cleaned IDs
valid? Print the result.

Then extract just the **numeric part** from each valid ID using a capture group:
e.g., `'P001'` → `1`, `'P019'` → `19`.

Print the list of extracted numbers.

*Hint: `re.match(r'^P(\d{3})$', pid)` - the parentheses create a capture group
so `match.group(1)` gives you just the 3 digits.*


````python
# Task 3.2 - Validate patient ID format
print(f'CleanIDs:{clean_ids}')  #lets see what is in clean ids to make a pattern
#start 1 P and 3 digits End
id_pattern = re.compile(r'^P(\d{3})$')

valid_ids   = []
invalid_ids = []
id_numbers  = []   # the numeric portion only

for pid in clean_ids:
    m = id_pattern.match(pid)
    # --- your validation and extraction code here ---
    if m:
        valid_ids.append(pid)
        id_numbers.append(int(m.group(1)))
    else:
        invalid_ids.append(pid)

print("All IDs valid:", len(invalid_ids) == 0)
print("Invalid IDs:", invalid_ids)
print("ID numbers extracted:", id_numbers)
````

### Task 3.3 - Extract and Count Date Formats `[4 marks]`

Dates arrive in three formats:
- `2024-01-15` - ISO format (YYYY-MM-DD)
- `15/01/2024` - DD/MM/YYYY
- `Jan 16 2024` - Month-name DD YYYY

**Part A:** Use `re.findall()` with three separate patterns to count how many
records use each format. Print the counts.

**Part B:** Extract the **year** from every admission date using one flexible
pattern. The year is always 4 consecutive digits. Build `admission_years` - a
list of year strings extracted from all 20 records.

Print `admission_years` and confirm that all records are from the same year.

*Hint for Part A:*
- ISO: `r'\d{4}-\d{2}-\d{2}'`
- DD/MM: `r'\d{2}/\d{2}/\d{4}'`
- Month name: `r'[A-Za-z]{3}\s+\d{1,2}\s+\d{4}'`


````python
# Task 3.3 - Extract and count date formats
iso_pattern   = re.compile(r'\d{4}-\d{2}-\d{2}')
ddmm_pattern  = re.compile(r'\d{2}/\d{2}/\d{4}')
month_pattern = re.compile(r'[A-Za-z]{3}\s+\d{1,2}\s+\d{4}')
year_pattern  = re.compile(r'\d{4}')

iso_count   = 0
ddmm_count  = 0
month_count = 0
admission_years = []

for record in records:
    fields = record.split(',')
    date   = fields[8].strip()
    # --- Part A: count formats ---
    if iso_pattern.findall(date):
        iso_count += 1
    elif ddmm_pattern.findall(date):
        ddmm_count += 1
    elif month_pattern.findall(date):
        month_count += 1

    # --- Part B: extract year ---
    year_match = year_pattern.search(date)
    admission_years.append(year_match.group())

print(f"ISO format (YYYY-MM-DD): {iso_count}")
print(f"DD/MM/YYYY format      : {ddmm_count}")
print(f"Month-name format      : {month_count}")

print(f"\nAll years: {admission_years}")
print(f"All from same year: {len(set(admission_years)) == 1}")
````

### Task 4.3 - Build Patient Summary Tuples `[3 marks]`

Create a list called `patient_records` where each element is an **immutable

  go at the end.

## Bonus Task - Patient Risk Flagging System `[+10 marks]`

The medical team wants to know which patients need urgent attention.
A patient is flagged as **HIGH RISK** if they meet **all three** of:
- Temperature ≥ 38.5°C
- Age ≥ 60 years old
- Valid phone number (so they can be contacted)

A patient is flagged as **MODERATE RISK** if they meet **any two** of the above.

All others are **LOW RISK**.

**Your task:**

Build a list called `risk_report` where each element is a tuple:
`(patient_id, name, risk_level)`

Use a `for` loop, `if/elif/else`, and the data you already have.

Print a formatted risk report grouped by level:
- HIGH RISK patients first
- Then MODERATE RISK
- Then LOW RISK

Also print a one-line summary: how many patients are in each category.

*This task tests whether you can combine everything - data from multiple
cleaned lists, conditional logic, and formatted output - into a coherent result.*


````python
# Bonus - Patient Risk Flagging System
risk_report = []

for i in range(len(records)):
    pid      = clean_ids[i]
    name     = clean_names[i]
    age      = clean_ages[i]
    temp     = clean_temps[i]
    phone_ok = clean_ids[i] not in invalid_records  # True if phone is valid

    # --- Your risk classification logic here ---
    # Count how many risk criteria are met:
    #   criteria 1: temp >= 38.5
    #   criteria 2: age >= 60 (careful: age might be None)
    #   criteria 3: phone is valid
    criteria_met = 0
    if temp >= 38.5:
        criteria_met += 1
    if age is not None and age >= 60:
        criteria_met += 1
    if phone_ok:
        criteria_met += 1

    # If all 3 met   -> HIGH RISK
    # If any 2 met   -> MODERATE RISK
    # Otherwise      -> LOW RISK
    if criteria_met == 3:
        risk_level = "HIGH RISK"
    elif criteria_met == 2:
        risk_level = "MODERATE RISK"
    else:
        risk_level = "LOW RISK"

    risk_report.append((pid, name, risk_level))

# Print grouped report
print("=" * 55)
print("   PATIENT RISK REPORT")
print("=" * 55)

for level in ["HIGH RISK", "MODERATE RISK", "LOW RISK"]:
    patients_at_level = [r for r in risk_report if r[2] == level]
    print(f"\n{level} ({len(patients_at_level)} patients):")
    print("─" * 40)
    for pid, name, _ in patients_at_level:
        print(f"  {pid}  {name}")

high_count     = len([r for r in risk_report if r[2] == "HIGH RISK"])
moderate_count = len([r for r in risk_report if r[2] == "MODERATE RISK"])
low_count      = len([r for r in risk_report if r[2] == "LOW RISK"])
print(f"\nSummary: HIGH={high_count}  MODERATE={moderate_count}  LOW={low_count}")
````
