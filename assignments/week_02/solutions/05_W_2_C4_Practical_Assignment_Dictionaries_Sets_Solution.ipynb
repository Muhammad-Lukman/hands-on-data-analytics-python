### Q1 - Build a lookup dict from two parallel lists `[3 marks]`

You are given two parallel lists: `cities` and `populations` (in millions).
Build a dictionary `city_pop` that maps each city to its population.

Then:
- Print the full dictionary
- Look up Lahore's population using `.get()`
- Look up a city that does **not** exist (`"Quetta"`) using `.get()` with
  the default string `"not in database"`

```
cities      = ["Karachi", "Lahore", "Islamabad", "Faisalabad", "Rawalpindi"]
populations = [16.1, 13.1, 1.2, 3.6, 2.3]
```

*Hint: `dict(zip(....))` pairs elements by position.*

### Q2 - Mutate a dictionary: add, update, and remove `[3 marks]`

Start with this patient dictionary:
```python
patient = {"id": "P0042", "name": "Ahmad Raza", "age": 34, "temp": 38.7}
```

Perform these four operations **in order**, printing the dictionary after each:
1. Add a new key `"blood_group"` with value `"A+"`
2. Update `"age"` to `35`
3. Use `.setdefault()` to add `"status": "admitted"` - then call
   `.setdefault("age", 99)` and explain in a comment why `"age"` does **not** become 99
4. Remove `"temp"` using `.pop()` and store the removed value - print it


````python
patient = {"id": "P0042", "name": "Ahmad Raza", "age": 34, "temp": 38.7}

# 1. Add blood_group
patient["blood_group"] = "A+"
print(patient)

# 2. Update age
patient["age"] = 35
print(patient)

# 3. setdefault
patient.setdefault("status", "admitted")
patient.setdefault("age", 99)
# "age" does NOT become 99 because setdefault() only inserts the given
# default value when the key is missing. Since "age" already exists (35),
# setdefault() just returns the existing value and leaves it untouched.
print(patient)

# 4. pop temp
removed_temp = patient.pop("temp")
print(patient)
print("Removed temp value:", removed_temp)
````

### Q4 - Dictionary comprehension: transform and filter `[3 marks]`

You have a dict of student scores:
```python
scores = {"Ahmad": 88, "Sara": 42, "Bilal": 76, "Zara": 55,
          "Hassan": 91, "Nida": 38, "Usman": 63}
```

Using **dict comprehensions** (one comprehension per task):
1. Build `grade_dict` that maps each name to `"Pass"` if score >= 50,
   else `"Fail"`
2. Build `high_scorers` containing only students who scored **above 75**,
   with their scores
3. Build `scaled` where every score is normalised to 0-100 by dividing
   by the max score - round to 2 decimal places

Print all three dicts.


````python
scores = {"Ahmad": 88, "Sara": 42, "Bilal": 76, "Zara": 55,
          "Hassan": 91, "Nida": 38, "Usman": 63}

# 1. grade_dict
grade_dict = {name: ("Pass" if score >= 50 else "Fail") for name, score in scores.items()}

# 2. high_scorers
high_scorers = {name: score for name, score in scores.items() if score > 75}

# 3. scaled
max_score = max(scores.values())
scaled = {name: round(score / max_score * 100, 2) for name, score in scores.items()}

print("Grade dict:", grade_dict)
print("High scorers:", high_scorers)
print("Scaled scores:", scaled)
````

### Q5 - Frequency counter using dict comprehension `[3 marks]`

Given this list of diagnoses from a clinic:
```python
diagnoses = ["Fever", "Diabetes", "Fever", "Hypertension", "Diabetes",
             "Fever", "Asthma", "Hypertension", "Fever", "Diabetes",
             "Asthma", "Fever", "Hypertension", "Diabetes", "Fever"]
```

1. Build a frequency dictionary `freq` using a dict comprehension.
   Use `set()` to get unique diagnoses and `.count()` to tally each.
2. Print each diagnosis and its count, sorted by count from **highest to lowest**.
3. Find and print the most common diagnosis using `max(freq, key=freq.get)`.

*Hint: `sorted(freq, key=freq.get, reverse=True)` gives keys sorted by value.*


````python
diagnoses = ["Fever", "Diabetes", "Fever", "Hypertension", "Diabetes",
             "Fever", "Asthma", "Hypertension", "Fever", "Diabetes",
             "Asthma", "Fever", "Hypertension", "Diabetes", "Fever"]

# your code here
# 1. Frequency dict
freq = {diagnosis: diagnoses.count(diagnosis) for diagnosis in set(diagnoses)}

# 2. Print sorted by count, highest to lowest
for diagnosis in sorted(freq, key=freq.get, reverse=True):
    print(f"{diagnosis}: {freq[diagnosis]}")

# 3. Most common diagnosis
most_common = max(freq, key=freq.get)
print("Most common diagnosis:", most_common)
````

### Q8 - .issubset(), .issuperset(), .isdisjoint() `[3 marks]`

```python
core_drugs     = {"Paracetamol", "Amoxicillin", "Omeprazole"}
pharmacy_stock = {"Paracetamol", "Amoxicillin", "Omeprazole",
                  "Metformin", "Atorvastatin", "Lisinopril"}
banned_drugs   = {"Thalidomide", "Phenacetin", "Aminopyrine"}
```

Answer these questions using the appropriate set methods and print
a clear `True/False` answer with a one-line explanation for each:

1. Does the pharmacy have all the core drugs in stock?
2. Does the pharmacy's stock fully contain the core drug list?
3. Is there any overlap between `pharmacy_stock` and `banned_drugs`?
   (This is a compliance check - no overlap is the safe answer.)


````python
core_drugs     = {"Paracetamol", "Amoxicillin", "Omeprazole"}
pharmacy_stock = {"Paracetamol", "Amoxicillin", "Omeprazole",
                  "Metformin", "Atorvastatin", "Lisinopril"}
banned_drugs   = {"Thalidomide", "Phenacetin", "Aminopyrine"}

# 1. Does the pharmacy have all the core drugs in stock?
has_all_core = core_drugs.issubset(pharmacy_stock)
print(f"{has_all_core} - pharmacy stock contains every core drug")

# 2. Does the pharmacy's stock fully contain the core drug list?
fully_contains = pharmacy_stock.issuperset(core_drugs)
print(f"{fully_contains} - pharmacy_stock is a superset of core_drugs")

# 3. Is there any overlap between pharmacy_stock and banned_drugs?
no_overlap = pharmacy_stock.isdisjoint(banned_drugs)
print(f"{no_overlap} - pharmacy_stock and banned_drugs share no items (disjoint means safe)")
````

### Q9 - Set comprehension `[3 marks]`

```python
product_prices = {"Laptop": 85000, "Mouse": 1200, "Keyboard": 2500,
                  "Monitor": 32000, "Webcam": 4500, "Headset": 6800,
                  "USB Hub": 1800, "Desk Lamp": 950, "Mousepad": 450}
```

Using **set comprehensions**:
1. Build a set of all product names that cost **more than 5000**
2. Build a set of the **first letter** of every product name
   (this shows the alphabet coverage of your catalog)
3. Build a set of all price values that are **divisible by 1000**
   (these are round-number prices)

Print all three sets.


````python
product_prices = {"Laptop": 85000, "Mouse": 1200, "Keyboard": 2500,
                  "Monitor": 32000, "Webcam": 4500, "Headset": 6800,
                  "USB Hub": 1800, "Desk Lamp": 950, "Mousepad": 450}

# 1. expensive items set
expensive_items = {name for name, price in product_prices.items() if price > 5000}

# 2. first letters set
first_letters = {name[0] for name in product_prices}

# 3. round prices set
round_prices = {price for price in product_prices.values() if price % 1000 == 0}

print("Expensive items (>5000):", expensive_items)
print("First letters:", first_letters)
print("Round-number prices:", round_prices)
````

### Q10 - Combine dicts and sets `[3 marks]`

This question requires you to use both structures together.

```python
team_a_skills = {"Ahmad": {"Python", "SQL", "Excel"},
                 "Sara":  {"Python", "R", "Tableau"},
                 "Bilal": {"Java", "SQL", "Power BI"}}

team_b_skills = {"Zara":   {"Python", "SQL", "Spark"},
                 "Hassan": {"Python", "Machine Learning", "SQL"},
                 "Nida":   {"R", "Tableau", "Excel"}}
```

1. Build `all_team_a_skills`: the **union** of all skill sets in team A
   (combine every person's skills into one big set)
2. Build `all_team_b_skills`: the same for team B
3. Print the skills that **appear in both teams** (intersection)
4. Print skills that **only team A** has
5. Print skills that **only team B** has
6. Build a dict `skill_count` that maps each unique skill (across both teams)
   to the total number of people (across both teams) who have it.
   Use a dict comprehension.

*Hint for 1: use a for loop or nested comprehension to union all sets together.*


````python
team_a_skills = {"Ahmad": {"Python", "SQL", "Excel"},
                 "Sara":  {"Python", "R", "Tableau"},
                 "Bilal": {"Java", "SQL", "Power BI"}}

team_b_skills = {"Zara":   {"Python", "SQL", "Spark"},
                 "Hassan": {"Python", "Machine Learning", "SQL"},
                 "Nida":   {"R", "Tableau", "Excel"}}

# 1. all team A skills
all_team_a_skills = set()
for skills in team_a_skills.values():
    all_team_a_skills |= skills
print("All team A skills:", all_team_a_skills)

# 2. all team B skills
all_team_b_skills = set()
for skills in team_b_skills.values():
    all_team_b_skills |= skills
print("All team B skills:", all_team_b_skills)

# 3. shared skills
shared_skills = all_team_a_skills & all_team_b_skills
print("Shared skills:", shared_skills)

# 4. only team A
only_team_a = all_team_a_skills - all_team_b_skills
print("Only team A:", only_team_a)

# 5. only team B
only_team_b = all_team_b_skills - all_team_a_skills
print("Only team B:", only_team_b)

# 6. skill_count dict
all_people_skills = {**team_a_skills, **team_b_skills}
all_skills = all_team_a_skills | all_team_b_skills
skill_count = {skill: sum(skill in person_skills for person_skills in all_people_skills.values())
               for skill in all_skills}
print("Skill count:", skill_count)
````

The logistics team needs to know what is stored in each warehouse.

   {"KHI-01": ["SKU-001", "SKU-005", "SKU-006"], "LHR-02": [...], ...}

````python
warehouse_skus = {}

# 1. Build the inverted index
for sku, details in product_catalog.items():
    warehouse = details["inventory"]["warehouse"]
    warehouse_skus.setdefault(warehouse, []).append(sku)

# 2. Print per-warehouse summary
for warehouse, skus in warehouse_skus.items():
    total_units = sum(product_catalog[sku]["inventory"]["in_stock"] for sku in skus)
    print(f"{warehouse}: {skus} | Total in-stock units: {total_units}")

# 3. Highest stock value warehouse
best_warehouse = max(
    warehouse_skus,
    key=lambda wh: sum(product_catalog[sku]["inventory"]["in_stock"] * product_catalog[sku]["pricing"]["cost_price"]
                        for sku in warehouse_skus[wh])
)
print(f"\nWarehouse with highest stock value: {best_warehouse}")
````

### Q27 - Build an inverted index: course to students `[3 marks]`

Build `course_roster`: a dict where each **course code** maps to a
**list of student names** enrolled in that course.

```python
{"CS-301": ["Ahmad Raza", "Sara Khan", "Bilal Ahmed", ...], "CS-302": [...], ...}
```

1. Build this using a for loop over all students and their `enrolled_courses`.
2. Print each course and its roster, sorted alphabetically by course code.
3. Find the course with the **most students** enrolled.
4. Find courses with **only one student** enrolled.


````python
course_roster = {}

# 1. Build course_roster
for details in student_registry.values():
    for course in details["enrolled_courses"]:
        course_roster.setdefault(course, []).append(details["name"])

# 2. Print sorted roster
print("--- Course Roster ---")
for course in sorted(course_roster):
    print(f"{course}: {course_roster[course]}")

# 3. Most popular course
most_popular_course = max(course_roster, key=lambda c: len(course_roster[c]))
print(f"\nMost popular course: {most_popular_course} ({len(course_roster[most_popular_course])} students)")

# 4. Courses with only one student
single_student_courses = [course for course, names in course_roster.items() if len(names) == 1]
print("Courses with only one student enrolled:", single_student_courses)
````

### Q28 - Club membership analysis `[3 marks]`

1. Build `club_members`: a dict mapping each **club name** to a
   **list of student names** who are members.

2. Print each club and its member count, sorted from **largest to smallest**.

3. Find students who are members of **more than 1 club**. Print their names
   and the number of clubs they belong to.

4. Find students with **no club memberships** at all (empty `clubs` list).
   Print their names.

5. Which club has the most members? Print the club name and member list.


````python
club_members = {}

# 1. Build club_members
for details in student_registry.values():
    for club in details["clubs"]:
        club_members.setdefault(club, []).append(details["name"])

# 2. Print sorted by size
print("--- Club Membership (largest to smallest) ---")
for club, members in sorted(club_members.items(), key=lambda x: len(x[1]), reverse=True):
    print(f"{club}: {len(members)} members")

# 3. Students in more than 1 club
print("\nStudents in more than 1 club:")
for details in student_registry.values():
    if len(details["clubs"]) > 1:
        print(f"{details['name']}: {len(details['clubs'])} clubs")

# 4. Students with no clubs
print("\nStudents with no club memberships:")
for details in student_registry.values():
    if len(details["clubs"]) == 0:
        print(details["name"])

# 5. Largest club
largest_club = max(club_members, key=lambda c: len(club_members[c]))
print(f"\nLargest club: {largest_club} - Members: {club_members[largest_club]}")
````

### Q29 - Cross-reference: skills vs grades `[3 marks]`

This question combines dicts, nested lists, and set operations.

1. Build `skill_to_students`: a dict mapping each skill to a list of
   student names who have that skill (similar to the inverted index in Q27,
   but for skills).

2. Find the **average grade** for students who know **Python** vs students
   who do **not** know Python. Is there a difference?
   (Average grade = mean of all values in `current_grades`.)

3. Build a dict `student_profile` for each student that contains:
   - `"name"`: the student's name
   - `"avg_grade"`: their average current grade (rounded to 1 dp)
   - `"avg_attendance"`: their average attendance (rounded to 1 dp)
   - `"skill_count"`: how many skills they have
   - `"status"`: `"Excellent"` if avg_grade >= 85 and avg_attendance >= 90,
     `"Good"` if avg_grade >= 70, else `"Needs Support"`

   Print the profile for every student.


````python
# 1. skill_to_students inverted index
skill_to_students = {}
for details in student_registry.values():
    for skill in details["skills"]:
        skill_to_students.setdefault(skill, []).append(details["name"])
print("Skill -> Students:", skill_to_students)

# 2. Python vs non-Python average grade
python_grades = [g for details in student_registry.values() if "Python" in details["skills"]
                  for g in details["current_grades"]]
non_python_grades = [g for details in student_registry.values() if "Python" not in details["skills"]
                      for g in details["current_grades"]]
avg_python = sum(python_grades) / len(python_grades) if python_grades else 0
avg_non_python = sum(non_python_grades) / len(non_python_grades) if non_python_grades else 0
print(f"\nAverage grade (knows Python): {avg_python:.2f}")
print(f"Average grade (no Python): {avg_non_python:.2f}")

# 3. student_profile dict
print("\n--- Student Profiles ---")
for details in student_registry.values():
    avg_grade = round(sum(details["current_grades"]) / len(details["current_grades"]), 1)
    avg_attendance = round(sum(details["attendance_pct"]) / len(details["attendance_pct"]), 1)
    skill_count = len(details["skills"])
    if avg_grade >= 85 and avg_attendance >= 90:
        status = "Excellent"
    elif avg_grade >= 70:
        status = "Good"
    else:
        status = "Needs Support"

    student_profile = {
        "name": details["name"],
        "avg_grade": avg_grade,
        "avg_attendance": avg_attendance,
        "skill_count": skill_count,
        "status": status
    }
    print(student_profile)
````

### Q30 - Full Department Report `[3 marks]`

Generate a comprehensive academic report for the department head.
Use everything you have computed in Q21-Q29.

Your report must include:

1. **Department Overview**
   - Total students, male count, female count
   - Highest CGPA student and lowest CGPA student

2. **Academic Performance**
   - Class average of current grades (average all individual grades)
   - Number of students with average grade >= 80 (Honours)
   - Number of students at academic risk (any grade < 60)

3. **Attendance Summary**
   - Students with average attendance >= 90% (Excellent)
   - Students with average attendance < 75% (At Risk)

4. **Skills Landscape**
   - Top 5 most common skills across the department
   - Skills held by fewer than 3 students (rare skills)

5. **Course Load**
   - Most popular currently enrolled course
   - Student with the most total courses (enrolled + completed combined)

Format the report using f-strings. Make it look like something you would
actually submit to a department chair.


````python
print("=" * 60)
print("DEPARTMENT OF COMPUTER SCIENCE - UAF")
print("SEMESTER 5 ACADEMIC REPORT")
print("=" * 60)

# 1. Department Overview
total_students = len(student_registry)
male_count = sum(1 for d in student_registry.values() if d["gender"] == "M")
female_count = sum(1 for d in student_registry.values() if d["gender"] == "F")
highest_cgpa_sid = max(student_registry, key=lambda sid: student_registry[sid]["cgpa"])
lowest_cgpa_sid = min(student_registry, key=lambda sid: student_registry[sid]["cgpa"])

print("\n1. DEPARTMENT OVERVIEW")
print(f"   Total students: {total_students} (Male: {male_count}, Female: {female_count})")
print(f"   Highest CGPA: {student_registry[highest_cgpa_sid]['name']} "
      f"({student_registry[highest_cgpa_sid]['cgpa']})")
print(f"   Lowest CGPA: {student_registry[lowest_cgpa_sid]['name']} "
      f"({student_registry[lowest_cgpa_sid]['cgpa']})")

# 2. Academic Performance
all_grades = [g for d in student_registry.values() for g in d["current_grades"]]
class_average = sum(all_grades) / len(all_grades)
honours_count = sum(1 for d in student_registry.values()
                     if sum(d["current_grades"]) / len(d["current_grades"]) >= 80)
at_risk_academic = sum(1 for d in student_registry.values() if any(g < 60 for g in d["current_grades"]))

print("\n2. ACADEMIC PERFORMANCE")
print(f"   Class average (all grades): {class_average:.2f}")
print(f"   Students with Honours (avg >= 80): {honours_count}")
print(f"   Students at academic risk (any grade < 60): {at_risk_academic}")

# 3. Attendance Summary
excellent_attendance = [d["name"] for d in student_registry.values()
                        if sum(d["attendance_pct"]) / len(d["attendance_pct"]) >= 90]
at_risk_attendance = [d["name"] for d in student_registry.values()
                       if sum(d["attendance_pct"]) / len(d["attendance_pct"]) < 75]

print("\n3. ATTENDANCE SUMMARY")
print(f"   Excellent attendance (>= 90%): {excellent_attendance}")
print(f"   At risk attendance (< 75%): {at_risk_attendance}")

# 4. Skills Landscape
all_skill_list = [skill for d in student_registry.values() for skill in d["skills"]]
unique_skill_set = set(all_skill_list)
skill_freq = {skill: all_skill_list.count(skill) for skill in unique_skill_set}
top_5_skills = sorted(skill_freq.items(), key=lambda x: x[1], reverse=True)[:5]
rare_skills = [skill for skill, count in skill_freq.items() if count < 3]

print("\n4. SKILLS LANDSCAPE")
print(f"   Top 5 skills: {top_5_skills}")
print(f"   Rare skills (< 3 students): {rare_skills}")

# 5. Course Load
course_freq = {}
for d in student_registry.values():
    for course in d["enrolled_courses"]:
        course_freq[course] = course_freq.get(course, 0) + 1
most_popular_course = max(course_freq, key=course_freq.get)
most_courses_sid = max(student_registry,
                        key=lambda sid: len(student_registry[sid]["enrolled_courses"])
                        + len(student_registry[sid]["completed_courses"]))

print("\n5. COURSE LOAD")
print(f"   Most popular course: {most_popular_course} ({course_freq[most_popular_course]} students)")
print(f"   Student with most total courses: {student_registry[most_courses_sid]['name']} "
      f"({len(student_registry[most_courses_sid]['enrolled_courses']) + len(student_registry[most_courses_sid]['completed_courses'])} courses)")

print("\n" + "=" * 60)
print("END OF REPORT")
print("=" * 60)
````
