### Q1 - Basic if/elif/else: air quality classifier `[3 marks]`

Write an `if/elif/else` chain that prints:

### Q3 - enumerate(): numbered city list `[3 marks]`


### Q4 - zip(): parallel list report `[3 marks]`

```python
fruits = ["Mango", "Apple", "Banana", "Guava"]
prices_per_kg = [180, 260, 90, 140]
in_stock = [True, True, False, True]
```

1. Use `zip()` in a `for` loop to print each fruit with its price, like
   `"Mango: PKR 180/kg"`.
2. Use `zip()` over **all three** lists together and print `"OUT OF STOCK"`
   next to any fruit where `in_stock` is `False`, otherwise print its price.
3. Add a fifth fruit, `"Kinnow"`, to `fruits` only (not the other two lists).
   Run your Q4.1 loop again unchanged - explain in a comment what happens
   and why `zip()` behaves this way.


````python
fruits = ["Mango", "Apple", "Banana", "Guava"]
prices_per_kg = [180, 260, 90, 140]
in_stock = [True, True, False, True]

# 1. fruit: price report
for fruit, price in zip(fruits, prices_per_kg):
    print(f"{fruit}: PKR {price}/kg")

print()

# 2. fruit: price or OUT OF STOCK
for fruit, price, stock in zip(fruits, prices_per_kg, in_stock):
    if stock:
        print(f"{fruit}: PKR {price}/kg")
    else:
        print(f"{fruit}: OUT OF STOCK")

print()

# 3. Add "Kinnow" to fruits, re-run loop 1, explain in a comment
fruits.append("Kinnow")
for fruit, price in zip(fruits, prices_per_kg):
    print(f"{fruit}: PKR {price}/kg")
# "Kinnow" never gets printed above, because zip() stops as soon as the
# SHORTEST input iterable is exhausted - prices_per_kg only has 4 values,
# so zip() produces only 4 pairs regardless of fruits having 5 items.
````

### Q5 - while loop with break: first divisible number `[3 marks]`

1. Using a `while` loop, find and print the **first number greater than 1000**
   that is divisible by both `7` and `11`. Start checking from `1001` upward
   and use `break` the moment you find it.
2. Using a **separate** `while` loop, simulate a countdown from `5` to `1`
   (print each number), then print `"Liftoff!"` after the loop ends -
   without using `break`.
3. In a comment, explain what would happen if you forgot to update the loop
   variable inside a `while` loop.


````python
# 1. First number > 1000 divisible by both 7 and 11
n = 1001
while True:
    if n % 7 == 0 and n % 11 == 0:
        print(f"First number > 1000 divisible by both 7 and 11: {n}")
        break
    n += 1

print()

# 2. Countdown from 5 to 1, then "Liftoff!"
count = 5
while count >= 1:
    print(count)
    count -= 1
print("Liftoff!")

# 3. Comment: what happens without updating the loop variable?
# If the loop variable is never updated inside the while loop, the
# condition never becomes False, so the loop runs forever (infinite loop)
# and the program hangs / has to be interrupted manually.
````

### Q9 - Nested if triage scoring `[3 marks]`

For **each** patient, compute a `score` using nested `if/elif` blocks
(same pattern as the lecture's `triage_score`, applied inline inside your
loop - not as a separate function):

- Temperature: `+3` if `>= 39.5`, `+2` if `>= 38.0`, `+1` if `>= 37.5`
- Blood pressure: `+3` if `>= 180`, `+2` if `>= 140`, `+1` if `>= 130`
- Age: `+1` if `>= 65`
- (Skip the glucose component for any patient whose glucose is `None`)

Convert each score to a level: `"CRITICAL"` (score >= 7), `"HIGH"` (>= 4),
`"MEDIUM"` (>= 2), else `"LOW"`.

Print a table of name, score, and level. Then print a count of how many
patients fall into each level (use a dictionary as a tally, updated with
`.get()` inside the loop).


````python
# your code here
level_counts = {}

print(f"{'Name':<15} {'Score':>6} {'Level':>10}")
print("-" * 34)

for p in patients:
    score = 0

    if p["temp"] >= 39.5:
        score += 3
    elif p["temp"] >= 38.0:
        score += 2
    elif p["temp"] >= 37.5:
        score += 1

    if p["bp"] >= 180:
        score += 3
    elif p["bp"] >= 140:
        score += 2
    elif p["bp"] >= 130:
        score += 1

    if p["age"] >= 65:
        score += 1

    if p["glucose"] is not None:
        if p["glucose"] >= 12.0:
            score += 2
        elif p["glucose"] >= 7.0:
            score += 1

    if score >= 7:
        level = "CRITICAL"
    elif score >= 4:
        level = "HIGH"
    elif score >= 2:
        level = "MEDIUM"
    else:
        level = "LOW"

    print(f"{p['name']:<15} {score:>6} {level:>10}")
    level_counts[level] = level_counts.get(level, 0) + 1

print("\nLevel counts:")
for level, count in level_counts.items():
    print(f"  {level}: {count}")
````

### Q10 - Ranked report with enumerate `[3 marks]`

1. Sort `patients` by `temp`, highest first (`sorted(..., key=..., reverse=True)`).
2. Use `enumerate()` (1-based) in a `for` loop to print a ranked list:
   `"1. Hassan Ali - 40.1C"`, `"2. Ahmad Raza - 39.6C"`, etc.
3. Using a `while` loop this time (not `for`), print only the **top 3**
   ranked patients from the same sorted list - stop the loop with a
   condition on a counter once you've printed 3.


````python
# 1 & 2. Sort by temp, print ranked list with enumerate
ranked = sorted(patients, key=lambda p: p["temp"], reverse=True)

for i, p in enumerate(ranked, start=1):
    print(f"{i}. {p['name']} - {p['temp']}C")

print()

# 3. Same top-3, but using a while loop
i = 0
while i < 3:
    p = ranked[i]
    print(f"{i + 1}. {p['name']} - {p['temp']}C")
    i += 1
````

````python
students = [
    {"name": "Ali Raza",     "weekly_attendance": [95, 90, 88, 92, 85, 80]},
    {"name": "Sara Khan",    "weekly_attendance": [100, 98, 95, 96, 99, 97]},
    {"name": "Ahmed Bhatti", "weekly_attendance": [70, 65, 60, 55, 50, 48]},
    {"name": "Mahnoor Ali",  "weekly_attendance": [88, 85, 90, 78, 72, 70]},
    {"name": "Bilal Hassan", "weekly_attendance": [60, 72, 68, 74, 65, 90]},
    {"name": "Iqra Yousaf",  "weekly_attendance": [92, 94, 91, 89, 93, 95]},
]
COURSES = ["Data Analytics", "Statistics", "Databases", "OOP", "Web Dev"]
course_names = ["Ali Raza", "Sara Khan", "Ahmed Bhatti", "Mahnoor Ali", "Bilal Hassan", "Iqra Yousaf"]
enrolled_course = ["Data Analytics", "Statistics", "Databases", "OOP", "Web Dev", "Data Analytics"]

print(f"Loaded {len(students)} students, {len(students[0]['weekly_attendance'])} weeks of data each.")
````

### Q11 - Nested loop: week-by-week status `[3 marks]`

Using a **nested** `for` loop (outer over students, inner over each
student's `weekly_attendance` with `enumerate()`), print a line for every
student-week combination:

```
Ali Raza - Week 1: 95% (Present)
Ali Raza - Week 2: 90% (Present)
...
```

### Q12 - while loop: first week below threshold `[3 marks]`

For **each** student (outer `for` loop), use an **inner `while` loop** to
find the **first week index** (0-based) where their attendance drops below
`75`. Print the student's name and that week number (1-based in your print
statement). If a student never drops below 75, print `"never dropped below
75%"` instead - your `while` loop should stop naturally (not need `break`)
once it has scanned all weeks for that student.


````python
# your code here
for s in students:
    week_idx = 0
    found = False
    while week_idx < len(s["weekly_attendance"]):
        if s["weekly_attendance"][week_idx] < 75:
            print(f"{s['name']}: first dropped below 75% in week {week_idx + 1}")
            found = True
            break
        week_idx += 1
    if not found:
        print(f"{s['name']}: never dropped below 75%")
````

### Q13 - continue: total absences excluding orientation week `[3 marks]`

Week 1 is an orientation week and should not count toward absences.

For each student, loop over their `weekly_attendance` with `enumerate()`.
Use `continue` to skip index `0` (week 1). For the remaining weeks, count
how many were `"Absent"` (`< 75`). Print each student's name and their
total absence count (weeks 2-6 only).


````python
# your code here
for s in students:
    absences = 0
    for week_idx, att in enumerate(s["weekly_attendance"]):
        if week_idx == 0:
            continue
        if att < 75:
            absences += 1
    print(f"{s['name']}: {absences} absence(s) in weeks 2-6")
````

### Q15 - Full attendance report `[3 marks]`

Generate a class attendance report using everything from Q11-Q14. Your
report must include, in order:

1. **Class Overview** - total students, total weeks tracked
2. **Attendance Health** - for each student, their average attendance
   (mean of all 6 weeks, rounded to 1 dp) and a status: `"Excellent"` if
   average `>= 90`, `"Good"` if `>= 75`, else `"At Risk"` - use `if/elif/else`
   inside a `for` loop
3. **Risk List** - print the names of every student whose status is
   `"At Risk"`, using `continue` to skip everyone else inside the loop
4. **Best Week** - loop over week indices (0-5) and, for each week, compute
   the class-wide average attendance for that week; print the week number
   with the **highest** class average (track it with a variable you update
   inside the loop, no need to sort)

Format the report with f-strings so it reads like something you'd actually
hand to a course coordinator.


````python
print("=" * 55)
print("WEEKLY ATTENDANCE REPORT - DATA ANALYTICS COHORT")
print("=" * 55)

# 1. Class Overview
num_weeks = len(students[0]["weekly_attendance"])
print(f"\nTotal students: {len(students)}")
print(f"Weeks tracked: {num_weeks}")

# 2. Attendance Health
print("\nAttendance Health:")
statuses = {}
for s in students:
    avg = round(sum(s["weekly_attendance"]) / len(s["weekly_attendance"]), 1)
    if avg >= 90:
        status = "Excellent"
    elif avg >= 75:
        status = "Good"
    else:
        status = "At Risk"
    statuses[s["name"]] = status
    print(f"  {s['name']:<15} avg={avg:>5.1f}%  status={status}")

# 3. Risk List
print("\nAt Risk students:")
for name, status in statuses.items():
    if status != "At Risk":
        continue
    print(f"  {name}")

# 4. Best Week
best_week = None
best_avg = -1
for week_idx in range(num_weeks):
    week_total = 0
    for s in students:
        week_total += s["weekly_attendance"][week_idx]
    week_avg = week_total / len(students)
    if week_avg > best_avg:
        best_avg = week_avg
        best_week = week_idx + 1

print(f"\nBest week: Week {best_week} (class average {best_avg:.1f}%)")
````
