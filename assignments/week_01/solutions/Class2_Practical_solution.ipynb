### Q17 - Employee ID Validation Using Regular Expressions

```python
import re

emp_id = "EMP-4587"
digit_found = re.search(r"\d", emp_id)
word_found = re.search(r"\w", emp_id)

print("Digit found:", bool(digit_found))
print("Alpha found:", bool(word_found))
```

### Q18 - Invoice Number Extraction

```python
report = "INV-1023 INV-4587 INV-9854 INV-1102"
invoices = re.findall(r"INV-\d+", report)

print("Extracted:", invoices)
print("Total count:", len(invoices))
```

### Q19 - Cleaning Phone Numbers

```python
numbers = "+92-300-1234567\n+92 321 9876543\n+92.333.7654321"
cleaned_numbers = re.sub(r"[-.\s]+", " ", numbers)

print("Original:\n", numbers)
print("\nCleaned:\n", cleaned_numbers)
```

### Q20 - Bank Transaction Parsing

```python
log = "TXN1001 Amount=4500 TXN1002 Amount=8200 TXN1003 Amount=15000"
txn_ids = re.findall(r"TXN\d+", log)
amounts = re.findall(r"Amount=(\d+)", log)

print("Transaction IDs:", txn_ids)
print("Amounts        :", amounts)
```

### Q21 - Social Media Hashtag Extraction

### Q27 - Strict Email Format Check Using Anchors

```python
r1 = "user@company.com"
r2 = "  user@company.com"
r3 = "user@company.com extra_text"

pattern = r"^\w+@\w+\.\w+$"

print("Record 1:", bool(re.search(pattern, r1)))
print("Record 2:", bool(re.search(pattern, r2)))
print("Record 3:", bool(re.search(pattern, r3)))
```
