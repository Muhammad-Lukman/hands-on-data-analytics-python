### Q5 - Views vs copies (1 mark)

```python
temps = np.array([36.5, 37.2, 38.9, 36.8, 39.1, 37.5])
```

**Part A - The view trap:**  
Create `high = temps[2:5]`. Modify `high[0] = 99.0`.  
Print `temps`. Explain in a comment why `temps` changed.

**Part B - Breaking the link:**  
Reset `temps` to its original values. Now create a **genuine copy** `safe = temps[2:5].copy()`.  
Modify `safe[0] = 99.0`. Print both `temps` and `safe`.  
Explain in a comment why `temps` is unchanged this time.

**Part C - Verification:**  
Use `np.shares_memory()` to confirm Part A shares memory and Part B does not.

````python
# Q5 - your answer here
temps = np.array([36.5, 37.2, 38.9, 36.8, 39.1, 37.5])

# Part A - view trap
high = temps[2:5]
print(f"\nhigh: {high} is just like a window to original array (temps).")
high[0] = 99.0
print("\nPart A - temps after modifying high(view):", temps)
# Slicing returns a VIEW sharing the same memory buffer,
# so writing to high[0] writes directly into temps[2].

# Part B - genuine copy
temps = np.array([36.5, 37.2, 38.9, 36.8, 39.1, 37.5])   # reset
safe = temps[2:5].copy()
safe[0] = 99.0
print("Part B - temps (unchanged):", temps)
print("Part B - safe (modified)  :", safe)
# .copy() allocates a new buffer; modifying safe never touches temps.

# Part C - verification
temps = np.array([36.5, 37.2, 38.9, 36.8, 39.1, 37.5])
high2 = temps[2:5]
safe2 = temps[2:5].copy()
print("Part C - high2 shares memory:", np.shares_memory(temps, high2))
print("Part C - safe2 shares memory:", np.shares_memory(temps, safe2))
````

### Q9 - Speed: NumPy vs Python loop (1 mark)

Compute the **sum of squares** of 5,000,000 numbers using:

**Method A - Python loop:**
```python
nums = list(range(1, 5_000_001))
```
Loop through `nums` and accumulate `total += x**2`.

**Method B - NumPy vectorised:**
```python
arr = np.arange(1, 5_000_001)
```
Compute `(arr**2).sum()` in a single expression.

Time both with `time.time()`. Print:
- Time for each method (in milliseconds, 2 decimal places (`round()`))
- The speedup factor (loop_time / numpy_time, integer)
- Confirm both methods give the same answer

````python
# Q9 - your answer here
import time

# Method A - Python loop
nums = list(range(1, 5_000_001))
t0 = time.time()
total_loop = 0
for x in nums:
    total_loop += x ** 2
t_loop = (time.time() - t0) * 1000

# Method B - NumPy vectorised
arr = np.arange(1, 5_000_001, dtype=object)     #numpy's int64 is different and specific as compare to python int, so we use dtype = object which is same as saying dtype = int to match python's int
t0 = time.time()
total_np = (arr ** 2).sum()
t_np = (time.time() - t0) * 1000

print(f"Python loop : {t_loop:.2f} ms  -->  result = {total_loop}")
print(f"NumPy       : {t_np:.2f} ms  -->  result = {total_np}")
print(f"Speedup     : {int(t_loop / t_np)}×")
print(f"Same answer : {total_loop == total_np}")
````

A pharmacokinetics study measures drug concentration every 0.5 hours

1. Create a `time_points` array using `np.linspace()` for hours 0.0 through 12.0  
   at 0.5-hour intervals (25 points). Print it.
