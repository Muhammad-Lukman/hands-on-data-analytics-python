## 1. String fundamentals

A string is an **immutable, ordered sequence of characters**:
- *Immutable* - you cannot change a character in place; every method call gives you back a brand-new string
- *Ordered* - every character sits at a fixed position (index)

````python
# Strings can be created with single, double, or triple quotes - all equivalent
s1 = 'Hello'
s2 = "Hello"
s3 = "It's fine here"      # double quotes let you embed apostrophes
s4 = 'She said "hi"'       # single quotes let you embed double quotes
print(s1, s2, s3, s4)

# Strings are IMMUTABLE - you cannot change a character in place
name = "ahmad"
try:
    name[0] = "A"
except TypeError as e:
    print(f"Cannot modify in place: {e}")

# To 'change' a string, you build a NEW one instead
name = "A" + name[1:]
print(f"New string: {name}")

# Operators we already know still work on strings
account = "PK-2026-778"
print(f"length      : {len(account)}")
print(f"membership  : {'2026' in account}")
print(f"repetition  : {'ab' * 3}")
print(f"concatenate : {'PK' + '-' + '2026'}")
````

## 2. Indexing

Each character has a position, counted two ways: **positive** from the left (starting at 0) and **negative** from the right (starting at -1).

````python
code_str = "COMSATS"
#            0123456    <- positive indices
#           -7-6-5-4-3-2-1  <- negative indices

print(f"code_str[0]  = {code_str[0]!r}")   # first character
print(f"code_str[3]  = {code_str[3]!r}")   # middle character
print(f"code_str[-1] = {code_str[-1]!r}")  # last character
print(f"code_str[-2] = {code_str[-2]!r}")  # second-last character

# indexing past the end raises an error - this is where exception handling helps
try:
    print(code_str[50])
except IndexError as e:
    print(f"IndexError: {e}")
````

## 3. Slicing

The `[start:stop:step]` mechanism is the same one you'll later use on Pandas Series and NumPy arrays - learning it once here pays off repeatedly. `stop` is always **exclusive**.

````python
seq = "ATCGATCG"

print("=== stop is EXCLUSIVE ===")
print(f"seq[0:3]  = {seq[0:3]!r}")
print(f"seq[:3]   = {seq[:3]!r}")     # start defaults to 0
print(f"seq[3:]   = {seq[3:]!r}")     # stop defaults to the end
print(f"seq[:]    = {seq[:]!r}")      # full copy of the string

print("\n=== step slicing ===")
print(f"seq[::2]  = {seq[::2]!r}")    # every 2nd character
print(f"seq[1::2] = {seq[1::2]!r}")   # every 2nd character, starting at index 1

print("\n=== reverse slicing ===")
print(f"seq[::-1]   = {seq[::-1]!r}")     # reverse the whole string
print(f"seq[5:1:-1] = {seq[5:1:-1]!r}")   # reverse a sub-range
````

## 4. Core string methods

Every method below returns a **new** string (remember: immutable), so you can chain them one after another.

````python
sentence = "Data Analytics is Fun, Data is Powerful"

print(f".upper() -> {sentence.upper()}")
print(f".lower() -> {sentence.lower()}")

# .split() breaks a string into a list, on a separator (,, tab, \n etc default: whitespace)
words = sentence.split()
print(f"\n.split()        -> {words}")

csv_row = "Ahmad,22,Faisalabad"
fields = csv_row.split(",")
print(f".split(',')     -> {fields}")

#.join() is the reverse of split - stitches a list back into one string
print(f".join()         -> {' '.join(words)}")
print(f".join() (dash)  -> {'-'.join(fields)}")

# .replace(old, new) swaps every occurrence of old with new
print(f"\n.replace()      -> {sentence.replace('Data', 'ML')}")

# .find() returns the first index, or -1 if the substring isn't there
print(f"\n.find('Fun')    -> {sentence.find('Fun')}")
print(f".find('Excel')  -> {sentence.find('Excel')}")

# .count() counts non-overlapping occurrences
print(f".count('Data')  -> {sentence.count('Data')}")
````

## 6. Formatting and alignment

`.ljust()`, `.rjust()`, and `.center()` pad a string to a fixed width - useful for lining up report output. f-strings can do the identical job inline.

````python
code = "CS101"
print(f"|{code.ljust(15)}|")        # left align, pad with spaces on the right
print(f"|{code.rjust(15)}|")        # right align, pad with spaces on the left
print(f"|{code.center(15)}|")       # center, pad both sides
print(f"|{code.center(15, '-')}|")  # center with a custom pad character

# f-string alignment does the exact same job without calling a method
print(f"|{code:<15}|")
print(f"|{code:>15}|")
print(f"|{code:^15}|")

#f-string precision - reviewed from Class 1, applied here to a computed value
gpa = 3.86142
print(f"\nGPA: {gpa:.2f}")   # 2 decimal places
````

## 8. Regular expressions

Today we use exactly three functions and four metacharacters - that's enough
to do real work. More regex (anchors, character sets, groups) comes later
once you have loops to put it to use.

| Function | What it does |
|---|---|
| `re.search(pattern, text)` | Finds the FIRST match anywhere in the text, or `None` |
| `re.findall(pattern, text)` | Returns ALL matches as a list |
| `re.sub(pattern, repl, text)` | Replaces every match with `repl` |

| Metacharacter | Matches |
|---|---|
| `\d` | Any digit 0–9 |
| `\w` | Any letter, digit, or underscore |
| `\s` | Any whitespace (space, tab, newline) |
| `*` | 0 or more of the previous character |
| `+` | 1 or more of the previous character |
| `?` | 0 or 1 of the previous character (optional) |
| `\D` | Non-digit |
| `.` | Any character except newline |
| `^` | Start of string |
| `$` | End of string |
| `{n}` | Exactly n times |
| `{n,m}` | Between n and m |
| `[ABC]` | Any char from set |
| `[^ABC]` | Any char NOT in set |
| `(...)` | Capture group |

````python
import re

text = "Order 2026 shipped on day 5 with 3 items"

# re.search() - finds the FIRST match anywhere in the string, or None
m = re.search(r"\d+", text)
print(f"first number found: {m.group()}")

# re.findall() - returns ALL matches as a list
all_numbers = re.findall(r"\d+", text)
print(f"all numbers: {all_numbers}")

# \w matches any letter, digit, or underscore
words_only = re.findall(r"\w+", text)   
print(f"all words: {words_only}")

# \s matches whitespace - useful for cleaning up messy spacing
messy = "Ahmad    Raza\t\tFaisalabad"
tidy = re.sub(r"\s+", "_", messy)
print(f"\nmessy -> {messy!r}")
print(f"tidy  -> {tidy!r}")

# quantifiers: * (0 or more), + (1 or more), ? (0 or 1)
print(f"\n're?' on 'r re ree' -> {re.findall(r're?', 'r re ree')}")
print(f"'re*' on 'r re ree' -> {re.findall(r're*', 'r re ree')}")
````

````python
'''
The symbols ? and * are quantifiers. They tell Python how many times it should look for the character right before them.
1. The ? Quantifier (0 or 1 occurrence)
    The pattern re? means: Match an 'r', followed by zero or one 'e'.
    It must find an r.
    If there is an e right after it, it grabs that too.
2. The * Quantifier (0 or more occurrences)
    The pattern re* means: Match an 'r', followed by zero or as many 'e's as possible.
    It must find an r.
    It will greedily swallow up every single continuous e that comes right after it.
    If there are more than one e, it ignores the extra ones because it only wants a maximum of one.

if used the + quantifier (re+), it would mean "an r followed by 1 or more es." 
It would completely skip the lone 'r' and output ['re', 'ree'])'''
````

### **Why regex over string methods???**
**Context:** A radiologist's report contains vital signs embedded in prose. Extract all numbers.

````python
report = (
    "Clinical Note: Patient presents with fever. Temperature recorded at 38.9 degrees\n"
    "Celsius. Blood pressure measured 145 over 92 mmHg. Heart rate 88 bpm. SpO2 97%%.\n"
    "Fasting glucose 7.4 mmol/L. BMI calculated at 28.3 kg/m2.\n"
    "Follow-up recommended in 14 days if temperature exceeds 38.5."
)

print("===Without regex: loop-based approach===")
words = report.split()
found_manual = []
for w in words:
    cleaned = w.strip(".,;:")
    try:
        found_manual.append(float(cleaned))
    except ValueError:
        pass
print(f"Values found: {found_manual}")

print("\n===With regex: one line===")
found_regex = re.findall(r"\d+\.?\d*", report)
print(f"Values found: {found_regex}")

print("\nExtract with context: parameter + value pairs")
pairs = re.findall(r"([A-Za-z]+(?:\s[A-Za-z]+)?(?:\s\d+)?[^:]*?)\s+(\d+\.?\d+)", report)
for param, val in pairs[:6]:
    print(f"  {param.strip():<30} {val}")
````

# Practical 1 - Cleaning a Raw Bank Statement Line

## Problem
A bank's legacy export system dumps each transaction as one messy string,
with inconsistent spacing and casing:

`raw_line = "   ahmad RAZA | acc: 011-778-42 | bal: rs 45230.7  "`

### Task
Using only string methods and slicing (no loops), display:
- The line with leading/trailing whitespace removed
- The customer name in Title Case
- The account number (the text between "acc:" and the next "|")
- The balance, formatted to exactly 2 decimal places

## Formula
`balance = float(balance_text)`, then display with `:.2f`

### Pseudocode
```
START
    READ raw_line
    STRIP leading and trailing whitespace -> stripped_line
    SPLIT stripped_line on "|" -> parts
    TAKE parts[0], STRIP it, apply .title() -> customer_name
    TAKE parts[1], REMOVE "acc:", STRIP -> account_number
    TAKE parts[2], REMOVE "bal:" and "rs", STRIP -> balance_text
    CONVERT balance_text to float -> balance
    DISPLAY customer_name, account_number, balance (2 decimals)
END
```

### Practical 6 - Why regex over string methods?
**Context:** A radiologist's report contains vital signs embedded in prose. Extract all numbers.

````python
# A radiologist's clinical note - numbers are embedded in prose
report = (
    "Clinical Note: Patient presents with fever. Temperature recorded at 38.9 degrees\n"
    "Celsius. Blood pressure measured 145 over 92 mmHg. Heart rate 88 bpm. SpO2 97%.\n"
    "Fasting glucose 7.4 mmol/L. BMI calculated at 28.3 kg/m2.\n"
    "Follow-up recommended in 14 days if temperature exceeds 38.5."
)

print("=== String method approach — .find() for specific known values ===")
# Works only if you know exactly what text appears before each value
temp_val = report[report.find("at ")+3 : report.find(" degrees")]
bp_start = report.find("measured ")+9
bp_val   = report[bp_start : bp_start+6].strip()
print(f"  Temperature extracted : '{temp_val}'")
print(f"  BP extracted          : '{bp_val}'")
print(f"  Problem: breaks the moment phrasing changes even slightly")

print("\n=== Regex approach — one pattern, all values ===")
all_numbers = re.findall(r"\d+\.?\d*", report)
print(f"  All numeric values: {all_numbers}")
print(f"  Count: {len(all_numbers)}")

print("\nRegex: parameter + value pairs")
pairs = re.findall(r"([A-Z][a-z]+(?:\s[a-z]+)?)\s+(\d+\.?\d*)", report)
print(f"  Extracted pairs: {pairs[:6]}")

print("\nKey lesson:")
print("  String methods -> fast and readable for FIXED, predictable structure")
print("  Regex          -> flexible and robust for VARIABLE, prose text")
````

````python
# Task 1: GC Content Calculation
# GC base pairs have 3 hydrogen bonds (stronger than AT's 2)
# GC content affects DNA melting temperature and sequencing difficulty

seq = "ATCGATCGATCGATCGGGGCCCC"
seq_clean = seq.upper().replace(" ", "")   # clean input first

g_count = seq_clean.count("G")
c_count = seq_clean.count("C")
a_count = seq_clean.count("A")
t_count = seq_clean.count("T")
total   = len(seq_clean)

gc_pct = (g_count + c_count) / total * 100
at_pct = (a_count + t_count) / total * 100

print(f"Sequence   : {seq_clean}")
print(f"Length     : {total} bp")
print()
print(f"A count    : {a_count}   ({a_count/total*100:.1f}%)")
print(f"T count    : {t_count}   ({t_count/total*100:.1f}%)")
print(f"G count    : {g_count}   ({g_count/total*100:.1f}%)")
print(f"C count    : {c_count}   ({c_count/total*100:.1f}%)")
print()
print(f"GC content : {gc_pct:.1f}%")
print(f"AT content : {at_pct:.1f}%")
print(f"GC + AT    : {gc_pct + at_pct:.1f}%  (should always be 100%)")

# Run on a second sequence to verify
seq2 = "GCGCGCGCGCGC"   # pure GC
g2 = seq2.count("G")
c2 = seq2.count("C")
print(f"\n'{seq2}' GC = {(g2+c2)/len(seq2)*100:.1f}%")
````

````python
# Task 2: Sequence Validation and Complement using str.maketrans()
dna_pattern = re.compile(r"^[ATCGatcg]+$")   # valid DNA characters only

# Five sequences to validate
s1 = "ATCGATCGATCG"
s2 = "ATUG"           # U = RNA base, not valid DNA
s3 = "ATCG123"        # digits not allowed
s4 = "atcgatcg"       # lowercase — valid DNA
s5 = "NNNATCG"        # N padding — invalid

print("Sequence Validation:")
print(f"{'Sequence':<20} {'Valid DNA?':<12} Note")
print("─" * 55)
print(f"  {s1:<18}  {str(bool(dna_pattern.match(s1))):<12} all ATCG")
print(f"  {s2:<18}  {str(bool(dna_pattern.match(s2))):<12} U is RNA not DNA")
print(f"  {s3:<18}  {str(bool(dna_pattern.match(s3))):<12} digits not allowed")
print(f"  {s4:<18}  {str(bool(dna_pattern.match(s4))):<12} lowercase ok")
print(f"  {s5:<18}  {str(bool(dna_pattern.match(s5))):<12} N not in [ATCGatcg]")

# Complement using str.maketrans() — a string method that builds a translation table
# A pairs with T,  T pairs with A,  C pairs with G,  G pairs with C
table = str.maketrans("ATCGatcg", "TAGCtagc")

rc1 = s1.translate(table)[::-1]   # complement then reverse = reverse complement
rc4 = s4.translate(table)[::-1]

print(f"\nReverse Complement:")
print(f"  '{s1}'  ->  '{rc1}'")
print(f"  '{s4}'  ->  '{rc4}'")

# Verify: reverse complement of reverse complement equals original
print(f"\nVerification (RevComp(RevComp) == original):")
print(f"  '{s1}' -> RevComp -> RevComp -> '{rc1.translate(table)[::-1]}'")
print(f"  Correct: {rc1.translate(table)[::-1] == s1}")
````

````python
# Task 3: Codon and Open Reading Frame (ORF) Analysis
# An ORF starts at ATG (start codon) and ends at a stop codon (TAA, TAG, or TGA)

genome = "ATGAAAGCCTTAGATCGATGATGCCCTAA"

print(f"Genome sequence: {genome}")
print(f"Length         : {len(genome)} bp")
print()

# Count codons using .count()
atg_count = genome.count("ATG")
taa_count = genome.count("TAA")
tag_count = genome.count("TAG")
tga_count = genome.count("TGA")
stop_total = taa_count + tag_count + tga_count

print(f"Start codon (ATG) count : {atg_count}")
print(f"Stop codons  TAA: {taa_count}  TAG: {tag_count}  TGA: {tga_count}")
print(f"Total stop codons       : {stop_total}")

# Find positions using .find()
first_atg  = genome.find("ATG")
first_stop = genome.find("TAA")    # first TAA stop codon

print(f"\nFirst ATG  at position: {first_atg}")
print(f"First TAA  at position: {first_stop}")

# Extract ORF — from first ATG to end of first stop codon
orf = genome[first_atg : first_stop + 3]
print(f"\nOpen Reading Frame: {orf}")
print(f"ORF length: {len(orf)} bp  ({len(orf)//3} codons including stop)")

# All ATG matches using re.findall
import re
atg_all   = re.findall(r"ATG", genome)
stops_all = re.findall(r"TAA|TAG|TGA", genome)
print(f"\nAll ATG matches  : {atg_all}")
print(f"All stop matches : {stops_all}")
````

## Summary

| Immutability | Every string method returns a NEW string - chain freely, modify never |

| `.strip()/.lstrip()/.rstrip()` | Remove characters from both/left/right ends |
| `.ljust()/.rjust()/.center()` | Pad a string to a fixed width |
| Raw strings `r"..."` | Backslashes are literal - always use for regex patterns |
| `re.search()` | First match anywhere; returns a match object or `None` |
| `re.findall()` | All matches, as a list |
| `re.sub()` | Pattern-based find-and-replace |
