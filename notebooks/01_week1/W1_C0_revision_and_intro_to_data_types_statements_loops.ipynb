### 2. Strings

Collection of one or more characters enclosed in single/double/triple quotes.
Multiline strings are printed using triple quotes.

````python
#basic declaration
sample_string = "Data_Science_2026"

#demonstrating multi-line structural outputs
medical_report = """
Clinical Notes Reference:
- Patient exhibits normal recovery markers.
- Blood pressure indexes verified stable.
"""
print(medical_report)

#verifying Immutability through structural assignment exception intercepts
try:
    sample_string[0] = "d"
except TypeError as error_message:
    print(f"Immutability Verification Triggered: {error_message}")
````

### 3. List

- An ordered sequence of items
- one of the most used data type and is very flexible (can store multiple data types)
- represented by square brackets `[]` where items are separated by `commas`
- Mutable

````python
# examples and list element accession and changing a value in a list
fruits = ["apple", "banana", "cherry"]
print(fruits, type(fruits))

# accessing elements
print("first item:", fruits[0])
print("last item:", fruits[-1])

# changing a value - lists are mutable
fruits[1] = "mango"
print("after change:", fruits)

# adding a new item - just a preview, covered in detail later
fruits.append("orange")
print("after append:", fruits)
````

### 4. Tuple

- An ordered sequence of items same as list
- Faster than list
- represented by parenthesis `()` where items are separated by `commas`
- Immutable


````python
# example with type check, element accession and changing a value
coordinates = (10, 20, 30)
print(coordinates, type(coordinates))

# accessing elements - same indexing as lists
print("first item:", coordinates[0])
print("last item:", coordinates[-1])

# tuples are immutable - this will fail
try:
    coordinates[0] = 99
except Exception as e:
    print(f"Cannot modify: {e}")
````

````python
coordinates[0] = 99
````

### 5. Dictionary
- An unordered collection of **key-value** pairs.
- Denoted by curly brackets `{}` where each item being a pair in the form of **key:value**.
- Enables highly efficient query response lookups across descriptive metadata labels.

````python
#example with type check and element accession and changing a value
student = {
    "name": "Ahmad",
    "age": 21,
    "program": "BSSE"
       }
print(student, type(student))

#accessing a value by its key
print("name:", student["name"])

#changing a value
student["age"] = 22
print("after change:", student)

#adding a new key-value pair
student["semester"] = 4
print("after adding:", student)
````

### 6. Sets
- An unordered collection of unique items.
- Each set element is unique (no duplicate).
- Highly effective for specialized logical operations (e.g., membership testing, unions, and intersections).
- Must be immutable
- denoted by curly brackets `{}`

````python
#example with type check and element accession
unique_marks = {85, 90, 85, 78, 90}   # duplicates are automatically removed
print(unique_marks, type(unique_marks))

# sets have no order and no indexing - this will fail
try:
    print(unique_marks[0])
except TypeError as e:
    print(f"Cannot index a set: {e}")

# membership check works though
print("is 90 in the set?", 90 in unique_marks)
````

## Conditional Statements
- `if` statement: Execute block if the statement evaluates to true.
- `if else` statement: Fallback alternative execution pathways when primary statements evaluate to false.
- `if else elif` statment: Multi-layered evaluation chains evaluating options sequentially.

### 1. if statement
```python
if [conditional expression]
    statement to execute
```

````python
# example of if statement
marks = 75

if marks >= 40:
    print("You passed!")
````

### 2. if else statement
```python
if [conditional expression]:
    statement to execute
else:
    alternative statement to execute
```

````python
# example of if-else statement
age = 16

if age >= 18:
    print("You can vote")
else:
    print("You cannot vote yet")
````

### 3. if elif else statement
```python
if [conditional expression]:
    statement to execute
elif [conditional expression]:
    statement to execute
else:
    alternative statement to execute
```

````python
# example of if-elif-else statement
marks = 82

if marks >= 90:
    grade = "A"
elif marks >= 75:
    grade = "B"
elif marks >= 60:
    grade = "C"
else:
    grade = "F"

print(f"Grade: {grade}")
````

## Loops
1. For loop
2. While Loop

### 1. For Loop
- To repeat data in code

- `range` function, by default it starts from zero and ends at n-1
- range parameters: `range(start, stop, step)`
- Reverse range: Using negative numbers to stride backwards through arrays safely

````python
#example of for loop with range and above data types

# range() - starts at 0, ends at n-1, by default
for i in range(5):
    print(i)

print()
# range with a start and a stop
for i in range(2, 6):
    print(i)

print()
# range with a start, stop, and step
for i in range(0, 10, 2):
    print(i)

print()
# reverse range - a negative step
for i in range(10, 0, -2):
    print(i)

print()
# looping over a list
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)

print()
# looping over a dictionary - gives you the keys by default
student = {"name": "Ahmad", "age": 21}
for key in student:
    print(key, ":", student[key])
````

### 2. While loop
- Infinite run till condition becomes False or broke intentionally with `break`
- Evaluates conditions continuously.

````python
#example of while loop with range and above data types

count = 0
while count < 5:
    print(count)
    count += 1

print()
# while loop with a break - runs "forever" until we choose to stop
n = 0
while True:
    if n == 3:
        break
    print(n)
    n += 1
````

````python
print("Conditional Break-Protected Loop Tracker")
safety_sentinel = 10
while True:
    print(f"System Scan Step: {safety_sentinel}")
    safety_sentinel -= 1
    if safety_sentinel == 7:
        print("Target Threshold Intercept matched. Triggering operational break safety cutoff mechanism.")
        break
````

## String Operations

- **string functions** (lower, upper, title, capitalize)
- **Indexing**: Target explicit positional byte frames (`[position]`).
- **Slicing**: Extract precise segments using `[start:stop:step]` parameters.
- **String iteration**: using for loop
- **Evaluation Predicates**: Validate text properties with structural checking rules (`isalpha()`, `isdigit()`, `isalnum()`).
- **Positional Methods**: Locate index coordinates using `.find()` and `.index()` methods.
- **Concatenation**

````python
#examples
text = "Data Analytics"

# indexing
print(text[0])       # first character
print(text[-1])      # last character

# slicing
print(text[0:4])      # 'Data'
print(text[5:])       # 'Analytics'
print(text[::-1])     # reversed

# string iteration
for ch in text[:4]:
    print(ch, end=" ")
print()

# string functions
print(text.lower())
print(text.upper())
print(text.title())
print(text.capitalize())

# find, index
print(text.find("Analytics"))    # returns an index, or -1 if not found
print(text.index("Analytics"))   # same idea, but raises an error if not found

# isalpha, isdigit, isalnum
print("Ahmad".isalpha())
print("2026".isdigit())
print("CS101".isalnum())

# concatenation
first = "Data"
second = "Analytics"
print(first + " " + second)
````
