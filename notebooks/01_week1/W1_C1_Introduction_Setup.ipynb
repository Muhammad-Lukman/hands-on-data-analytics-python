## 3. Variables and data types

Python uses **dynamic typing** - you don't declare a type. The value carries its own type. This feels looser than Java or C, but in data analytics it creates a specific class of bugs: a column that *looks* numeric but is actually a string of digits. Knowing your types saves hours of debugging.

| Type | Example | Used for |
|------|---------|----------|
| `int` | `34` | Whole numbers — age, count, patient ID |
| `float` | `37.2` | Decimal numbers — temperature, score, BMI |
| `str` | `"Sara Ahmed"` | Text — names, labels, DNA sequences |
| `bool` | `True` / `False` | Yes/No — pass/fail, present/absent |
| `NoneType` | `None` | Missing data — value not yet known |

These are not abstract — every real dataset you will analyse has columns of each type.

#### Variable naming rules

- First Character: Must begin with a letter (A-Z, a-z) or an underscore _. Never start a name with a number.
- Allowed Characters: Only contain alphanumeric characters (letters and numbers) and underscores.
- Special characters (e.g., @, -, !, %) are strictly prohibited.
- No Spaces: Whitespace cannot be used within a name.
- Case Sensitivity: Names are case-sensitive. For example, myVariable and myvariable are treated as two distinct variables.
- No Reserved Keywords: You cannot use built-in keywords reserved by the programming language for specific operations (e.g., if, for, print, int).

```python
# Variable naming rules and conventions
snake_case_var  = 42        # variables and functions: snake_case (all letters lowercase and words separated by underscores, easier to read)
MAX_SAMPLE_SIZE = 1000      # constants: UPPER_SNAKE_CASE
# PascalCase reserved for class names (covered later) - first letter of each word is capitalized typically reserved for class names

# Multiple assignment — assign different values in one line
x, y, z = 10, 20, 30
print(f"Before: x = {x},  y = {y},  z = {z}")

# Chained assignment — same value to multiple names at once
a = b = c = 0
print(f"a = {a},  b = {b},  c = {c}")

# Swap without a temp variable, many langs need a temp variable to swap i.e. in C temp = x; x = y; y = temp;
x, y = y, x
print(f"After swap: x = {x},  y = {y}")
```

## 4. Type inspection - `type()`, `isinstance()`, `id()`

Three functions that tell you *what* a variable is and *where* it lives in memory.

```python
# type() - tells you exactly what class a value belongs to
print(type(42))
print(type(3.14))    # <class 'float'>
print(type("hello"))
print(type(True))
print(type(None))

# .__name__ gives just the type name as a clean string
height = 1.75
score  = 95
print(f"\nheight is a {type(height).__name__}")
print(f"score  is a {type(score).__name__}")
```

```python
# isinstance() - answers "is this value of type T?" with True or False
# Preferred over type() for checking types in real code also accounts for inheritance

score = 88
print(isinstance(score, int))            # True
print(isinstance(score, float))          # False
print(isinstance(score, (int, float)))   # True - accepts a tuple of types to check against

# The most important isinstance fact: bool IS a subclass of int
# This is not a bug - it is designed this way intentionally
flag = True
print(f"\nIs flag a bool?  {isinstance(flag, bool)}")
print(f"Is flag an int?  {isinstance(flag, int)}")   # Also True!

# Why this matters: True acts like 1 and False like 0 in arithmetic
# we will use this extensively in NumPy to count flagged values
```

```python
# id() — returns the memory address of any Python object
# Two names with the same id() are literally the same object in memory

a = 5
b = 5
print(f"a = 5,  id(a) = {id(a)}")
print(f"b = 5,  id(b) = {id(b)}")
print(f"Same object (a is b): {a is b}")

# CPython caches small integers (-5 to 256) for performance.
# For numbers outside that range, each assignment usually creates a new object.
x = 1000
y = 1000
print(f"\nx = 1000,  id(x) = {id(x)}")
print(f"y = 1000,  id(y) = {id(y)}")
print(f"Same object (x is y): {x is y}")

# You will revisit this idea directly in the NumPy module:
# two array names can share the same underlying memory (a 'view'),
# and modifying one will change the other. id() is how you spot that.
# a = [1, 2, 3]
# b = a           # b is an alias — same object
# c = a.copy()    # c is a new object with same contents

# print(f"\nid(a) = {id(a)}")
# print(f"id(b) = {id(b)}  <- same as a")
# print(f"id(c) = {id(c)}  <- different object")

# b.append(99)
# print(f"\na after modifying b: {a}")  # a changed too - they share one object
# print(f"c unchanged:         {c}")
```

## 5. Arithmetic operators

Python has seven arithmetic operators. The two division operators are the most
important distinction: `/` always returns a float, `//` always returns an integer.

```python
a, b = 17, 5

print(f"  {a} + {b}   = {a + b}")      # addition
print(f"  {a} - {b}   = {a - b}")      # subtraction
print(f"  {a} * {b}   = {a * b}")      # multiplication
print(f"  {a} / {b}   = {a / b}")      # true division  - result is ALWAYS a float
print(f"  {a} // {b}  = {a // b}")     #floor division floor toward -infinity (minus infinity), rounds results down to the nearest whole number (towards -infinity)
print(f"  {a} % {b}   = {a % b}")      # modulo         - remainder after floor division
print(f"  {a} ** {b}  = {a ** b}")     # exponentiation - 17 to the power 5

#use: break a time interval into hours + leftover minutes
total_minutes = 137
hours         = total_minutes // 60    # how many complete hours?
leftover      = total_minutes % 60     # what's left over?
print(f"\n{total_minutes} minutes = {hours}h {leftover}m")

# The // subtlety
print(f"\n  7 // 2  = {7//2}")      # 3  (as expected)
print(f"  -7 // 2 = {-7//2}")       # -4  NOT -3 — floor is toward -infinity
```

## 6. Comparison operators

These always return `True` or `False`. They are the building blocks of every
filter condition you will write in Pandas and every mask you will apply in NumPy.

```python
temp = 38.9

print(f"Temperature: {temp} °C")
print(f"  > 37.5  (fever threshold)?  {temp > 37.5}")
print(f"  < 36.0  (hypothermia)?      {temp < 36.0}")
print(f"  == 37.0 (perfectly normal)? {temp == 37.0}")
print(f"  != 37.0 (not normal)?       {temp != 37.0}")
print(f"  >= 39.5 (high fever)?       {temp >= 39.5}")
print(f"  <= 37.5 (no fever)?         {temp <= 37.5}")

# Chaining comparisons works in Python (unlike most other languages)
print(f"\n36.0 <= temp <= 37.5 (normal range)? {36.0 <= temp <= 37.5}")
print(f"37.5 < temp < 39.5 (mild fever)?      {37.5 < temp < 39.5}")

# Float comparison trap
import math
print(f"\n  0.1 + 0.2 == 0.3      --->  {0.1+0.2==0.3}")   #False!
print(f"  0.1 + 0.2              =  {0.1+0.2}")
print(f"  math.isclose(...)      --->  {math.isclose(0.1+0.2, 0.3)}")  #True
```

## 7. Logical operators

`and`, `or`, and `not` combine boolean conditions into compound decisions.
Python evaluates them left to right and stops as soon as the result is determined.

```python
age       = 28
bmi       = 22.4
is_smoker = False

# 'and' - every condition must be True
healthy = age < 40 and bmi < 25 and not is_smoker
print(f"Healthy profile? {healthy}")

# 'or' - at least one condition must be True
has_diabetes     = False
has_hypertension = True
needs_monitoring = has_diabetes or has_hypertension
print(f"Needs monitoring? {needs_monitoring}")

# 'not' - flips True to False, False to True
print(f"Not a smoker? {not is_smoker}")

# Precedence: not > and > or
# When combining all three, use parentheses to make the intent unambiguous
blood_glucose = 8.5
is_at_risk = (blood_glucose > 7.0) and (not has_diabetes)
print(f"\nAt risk but undiagnosed? {is_at_risk}")

score, attendance = 82, 85
eligible = score >= 60 and attendance >= 75
print(f"\n  Eligible for certificate: {eligible}")
```

```python
A = False
B = True
result = A and B
print(f"result: {result}") #B never evaluated, AND returns True if both are true, if first is false, no need for furthur evaluation
A = True
result2 = A or B
print(f"result: {result2}") # B never evaluated, OR returns True if either is True so first is True answer is True
```

## 8. Augmented/Compound assignment operators

`x += 5` is shorthand for `x = x + 5`. The augmented form is cleaner,
slightly faster, and is what you will see in all professional code.

```python
balance = 1000.00
print(f"Starting balance: PKR {balance:,.2f}")

balance += 5000          # deposit
print(f"After deposit:    PKR {balance:,.2f}")

balance *= 1.05          # 5% annual return
print(f"After 5% return:  PKR {balance:,.2f}")

balance -= 150           # monthly maintenance fee
print(f"After fee:        PKR {balance:,.2f}")

balance /= 2             # split with partner
print(f"After split:      PKR {balance:,.2f}")

# Quick tour of all augmented operators on a simple counter
n = 100
print(f"\nStarting n = {n}")
n += 10;   print(f"  n += 10   --->  {n}")
n -= 5;    print(f"  n -= 5    --->  {n}")
n *= 2;    print(f"  n *= 2    --->  {n}")
n //= 3;   print(f"  n //= 3   --->  {n}")
n **= 2;   print(f"  n **= 2   --->  {n}")
n %= 50;   print(f"  n %= 50   --->  {n}")
```

```python
#precedence reminder (the order in which operations are performed in an expression): ** > unary - > * / // % > + - > comparisons (==, !=, >, <, >=, <=) > not > and > or
result = 2 + 3 * 4 ** 2
print(f"\n  2 + 3 * 4**2 = {result}")  # 50: 4**2=16, 3*16=48, 2+48=50
```

## 9. Membership and identity operators

`in` / `not in` — checks whether something appears inside a string (or later, a collection).
`is` / `is not` — checks whether two names refer to the *same object in memory*.

```python
# Membership - 'in' and 'not in' on strings
# (we will extend this to lists, sets, and dict-keys in later classes)
dna = "ATCGATGCTATCG"
print(f"Sequence: {dna}")
print(f"  'ATG' present?  {'ATG' in dna}")
print(f"  'TTT' present?  {'TTT' in dna}")
print(f"  'ATG' absent?   {'ATG' not in dna}")

email = "usman@uaf.edu.pk"
print(f"\nEmail: {email}")
print(f"  UAF address?    {'@uaf.edu.pk' in email}")
```

```python
# Identity - 'is' and 'is not'
# Checks object identity (if two variables point to the SAME object in memory), NOT value equality

# None is a singleton — there is exactly one None object in Python
result = None
print(f"result is None   : {result is None}")     # correct idiom for None checks
print(f"result == None   : {result == None}")      # works, but 'is None' is preferred

# The critical distinction: == compares values, 'is' compares identity
x = 257
y = 257
# y = x
print(f"\nx = 257, y = 257")
print(f"x == y  : {x == y}")      # True  - same value
print(f"x is y  : {x is y}")      # False - different objects in memory
# Rule for data analytics:
# Use ==   to compare values (99% of the time)
# Use is   only for None, True, or False checks

print(id(x), id(y))
```

## 10. Operator precedence

Python evaluates operators in a fixed order. Not knowing this order is the source
of subtle, hard-to-find bugs.

**Simplified order (highest to lowest):**
`**` ---> `* / // %` ---> `+ -` ---> comparisons ---> `not` ---> `and` ---> `or`

**Rule:** when in doubt, use parentheses. They cost nothing and prevent bugs.

```python
# The three most commonly misunderstood cases

# 1. Multiplication before addition — like school BODMAS
print(f"2 + 3 * 4     = {2 + 3 * 4}")       # 14, not 20
print(f"(2 + 3) * 4   = {(2 + 3) * 4}")     # 20 with parentheses

# 2. Exponentiation is right-associative (evaluated right to left)
print(f"\n2 ** 3 ** 2   = {2 ** 3 ** 2}")   # 512: 3**2=9 first, then 2**9
print(f"(2 ** 3) ** 2  = {(2 ** 3) ** 2}")  # 64: 2**3=8 first, then 8**2

# 3. The unary minus trap — the most common precedence bug
x = 5
print(f"\nx = {x}")
print(f"-x ** 2        = {-x ** 2}")        # -25! because -(x**2), not (-x)**2
print(f"(-x) ** 2      = {(-x) ** 2}")      # 25 — what most people intended

#example: order of operations in a real formula
score = 85
bonus = 5
adjusted = score + bonus * 2      # bonus is doubled, THEN added — is that the intent?
adjusted2 = (score + bonus) * 2   # or did you mean to double the whole sum?
print(f"\n85 + 5 * 2  = {adjusted}")
print(f"(85 + 5) * 2 = {adjusted2}")
```

## 11. Type conversion

Python does not silently convert types. When data arrives from a file, API, or
`input()`, it is almost always a string - you must convert it explicitly.

```python
# int() - string to integer (whole numbers only)
age_str = "22"
age     = int(age_str)
print(f"'{age_str}'  --->  {age}  (type: {type(age).__name__})")

# float() - string to float (accepts decimal strings)
temp_str = "38.9"
temp     = float(temp_str)
print(f"'{temp_str}' --> {temp}  (type: {type(temp).__name__})")

# str() - anything to string (always works)
patient_id = 1042
id_str     = str(patient_id)
print(f"{patient_id}    --->  '{id_str}'  (type: {type(id_str).__name__})")

# int() on a float DROPS the decimal — it does NOT round
temp_float = 38.9
print(f"\nint(38.9) = {int(temp_float)}  <-- truncates toward zero, never rounds")

# Two-step conversion when data looks like "3.7" and you need an int
value_str = "3.7"
value_int = int(float(value_str))   # float first, then int
print(f"int(float('3.7')) = {value_int}")
```

```python
# bool() - every value in Python is either truthy or falsy
# This is critical for analytics: counting True values = counting occurrences

# know what's falsy: 0, 0.0, "", [], {}, set(), None - bool function will return falls for these

print("Falsy values (bool() returns False):")
print(f"  bool(0)        = {bool(0)}")
print(f"  bool(0.0)      = {bool(0.0)}")
print(f"  bool('')       = {bool('')}")
print(f"  bool(None)     = {bool(None)}")

print("Truthy values (bool() returns True):")
print(f"  bool(1)        = {bool(1)}")
print(f"  bool(-5)       = {bool(-5)}")      # any non-zero number
print(f"  bool('hello')  = {bool('hello')}") # any non-empty string
print(f"  bool(0.001)    = {bool(0.001)}")

# Direct boolean arithmetic - works because bool is a subclass of int
fever   = True    # this patient has fever
normal1 = False
normal2 = False
fever2  = True

total_fever = fever + normal1 + normal2 + fever2   # True=1, False=0
print(f"\nFever count (addition of bools): {total_fever}")
```

### The float precision trap
**Context:** A hospital drug-dosing system computes `weight_kg * dose_per_kg`. Should the developer worry about float precision here?

```python
# Does this print True or False?
dose_per_kg = 0.1      # mg/kg
weight = 3.0            # kg

total_dose = dose_per_kg + dose_per_kg + dose_per_kg   # 3 doses for 1 kg
print(f"  3 x 0.1  =  {total_dose}")
print(f"  == 0.3?  --->  {total_dose == 0.3}")      #True?!!

# Why this matters in analytics:
# When you filter a Pandas column with df[df['value'] == 0.3]
# you may miss rows that are actually 0.30000000000000004
#to fix:
import math
print(f"  math.isclose() --->  {math.isclose(total_dose, 0.3)}")

# In data analytics: use round() for display, isclose() for comparison,
# and Decimal() when you need exact decimal arithmetic (banking, dosing)
```

```python
'''Computers store numbers in binary (base 2).
While whole numbers (integers) can be represented exactly in binary,
many decimal fractions (like 0.1, 0.2, 0.3) cannot be represented precisely in binary using a finite number of bits.
When you write 0.1 in your code, the computer stores the closest possible binary approximation of 0.1.
This approximation is ever so slightly larger or smaller than the true mathematical value.
When you add these approximated numbers together (0.1 + 0.1 + 0.1),
these tiny inaccuracies accumulate. The sum of three slightly-off approximations doesn't necessarily equal the exact approximation for 0.3'''
```

### Small integer caching and id()

```python
#  CPython pre-creates integer objects from -5 to 255 and reuses them.
# Assigning the same small number to two names gives you the SAME object.
# Above that range, each assignment may create a separate object.

a = 5
b = 5
print(f"a = 5,  b = 5")
print(f"id(a) = {id(a)},  id(b) = {id(b)}")
print(f"a is b (same object?) : {a is b}")         # True — cached

c = 1000
d = 1000
print(f"\nc = 1000,  d = 1000")
print(f"id(c) = {id(c)},  id(d) = {id(d)}")
print(f"c is d (same object?) : {c is d}")         # Often False
print(f"c == d (equal values?): {c == d}")         # Always True

# Lesson: == for comparing values. 'is' for checking identity.
# In data analytics you almost always want ==.
```

### The unary minus and exponentiation trap

```python
# ** binds more tightly than the - sign in front of a variable.
# -x**2 means -(x**2), not (-x)**2.
# This catches people writing physics or statistics formulas.

x = 5
print(f"x = {x}")
print(f"-x ** 2    = {-x**2}")       # -25 — NOT what most people expect
print(f"(-x) ** 2  = {(-x)**2}")     # 25  — explicit and unambiguous
print(f"-(x ** 2)  = {-(x**2)}")     # -25 — same as the first line, made visible

# The same trap in a real formula: variance uses x**2
mean = 3
val  = 7
deviation = val - mean

# Wrong: if you forgot parentheses around deviation
wrong  = -deviation**2
right  = -(deviation**2)
also   = (-deviation)**2     # different calculation entirely

print(f"\ndeviation = {deviation}")
print(f"-deviation**2    = {wrong}")    # negative — bug if you wanted positive
print(f"-(deviation**2)  = {right}")    # same
print(f"(-deviation)**2  = {also}")     # positive, but a different thing
```

### bool arithmetic (True is 1, False is 0)

```python
# Because bool is a subclass of int, you can do arithmetic with booleans.
# This is used intentionally in NumPy and Pandas to count True values.

print(f"True  + True   = {True + True}")     # 2
print(f"True  * 5      = {True * 5}")        # 5
print(f"False + 7      = {False + 7}")       # 7
print(f"True  / 2      = {True / 2}")        # 0.5
print(f"True  > False  = {True > False}")    # True  (1 > 0)
print(f"True  == 1     = {True == 1}")       # True
print(f"False == 0     = {False == 0}")      # True

# Practical pattern: count how many patients are flagged
# (We do this without a loop - pure arithmetic on boolean values)
fever1 = True
fever2 = False
fever3 = True
fever4 = True
fever5 = False

total_fever = fever1 + fever2 + fever3 + fever4 + fever5
total_patients = 5
print(f"\nFever count: {total_fever} out of {total_patients}")
print(f"Fever rate:  {total_fever / total_patients:.0%}")
```

### Floor division with negatives + floating-point precision

```python
# Two things that trip up almost every new Python programmer.

# Floor division ALWAYS rounds toward negative infinity - not toward zero
print("Floor division with negative numbers:")
print(f"  -7 // 2   = {-7 // 2}")     #  -4  (not -3!)
print(f"   7 // -2  = {7 // -2}")     #  -4  (same direction)
print(f"  -7 // -2  = {-7 // -2}")    #   3  (negative / negative)

# Floating-point precision - the most common numerical surprise
print("\nFloating-point precision:")
print(f"  0.1 + 0.2              = {0.1 + 0.2}")
print(f"  0.1 + 0.2 == 0.3       = {0.1 + 0.2 == 0.3}")  # False!
print(f"  0.1 + 0.2 - 0.3        = {0.1 + 0.2 - 0.3}")   # tiny non-zero error

# Why: 0.1 and 0.2 cannot be represented exactly in binary floating-point.
#for display: use round() or format with :.2f
result = 0.1 + 0.2
print(f"  rounded to 2 places:    {round(result, 2)}")
print(f"  formatted as .2f:       {result:.2f}")
print(f"  within tolerance?       {abs(result - 0.3) < 1e-10}")
```

## Practical Unit Converter

Build a unit converter using only what you have learned today.
**No functions. No loops. No `if` statements.**
Just variables, operators, expressions, and f-strings.

Change the input values at the top of the cell and re-run.
That is the only interactivity needed.

```python
#  INPUT VALUES — change these and re-run 
celsius      = 100.0       # temperature (°C)
weight_kg    = 72.0        # weight (kg)
distance_km  = 5.0         # distance (km)
feet         = 5           # height — feet component
inches       = 10          # height — inches component
pkr_amount   = 50000.0     # currency (Pakistani Rupees)

#  CONVERSION CONSTANTS 
LBS_PER_KG   = 2.20462
KG_PER_STONE = 6.35029
MI_PER_KM    = 0.621371
CM_PER_INCH  = 2.54
USD_RATE     = 0.00358      # 1 PKR = 0.00358 USD (approx.)
GBP_RATE     = 0.00282      # 1 PKR = 0.00282 GBP (approx.)

#  CALCULATIONS — sequential, one formula per line 

# 1. Temperature
fahrenheit = (celsius * 9 / 5) + 32
kelvin     = celsius + 273.15

# 2. Weight
pounds = weight_kg * LBS_PER_KG
stones = weight_kg / KG_PER_STONE

# 3. Distance
miles  = distance_km * MI_PER_KM
metres = distance_km * 1000

# 4. Height
total_inches = (feet * 12) + inches
total_cm     = total_inches * CM_PER_INCH

# 5. Currency
usd = pkr_amount * USD_RATE
gbp = pkr_amount * GBP_RATE

#  OUTPUT 
print("=" * 48)
print("            UNIT CONVERTER RESULTS")
print("=" * 48)

print(f"\nTemperature")
print(f"  {celsius} °C  -->  {fahrenheit:.1f} °F")
print(f"  {celsius} °C  -->  {kelvin:.2f} K")

print(f"\nWeight")
print(f"  {weight_kg} kg  -->  {pounds:.2f} lbs")
print(f"  {weight_kg} kg  -->  {stones:.2f} stone")

print(f"\nDistance")
print(f"  {distance_km} km  --> {miles:.3f} miles")
print(f"  {distance_km} km  -->  {metres:.0f} metres")

print(f"\nHeight")
print(f"  {feet}\' {inches}\"  {total_cm:.1f} cm")

print(f"\nCurrency")
print(f"  PKR {pkr_amount:,.0f}  -->  USD {usd:,.2f}")
print(f"  PKR {pkr_amount:,.0f}  -->  GBP {gbp:,.2f}")

print("\n" + "=" * 48)
print("No loops. No functions. All sequential.")
```

| `print(f"...")` | Always use f-strings for output - embed values with `{}` |

| `+=`, `-=`, `*=` ... | Augmented assignment - update in place, one line |
| `in`, `not in` | Membership check - works on strings now, collections later |
| `is`, `is not` | Identity check - almost exclusively for `None` comparisons |
| `int()`, `float()`, `str()`, `bool()` | Explicit conversion - required after every `input()` |
