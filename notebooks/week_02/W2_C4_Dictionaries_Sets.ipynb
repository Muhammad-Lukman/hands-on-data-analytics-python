## 1. Dictionary fundamentals

A dictionary stores **key ---> value** pairs. **Keys must be immutable (strings, numbers, tuples)**. Values can be anything. Order is preserved from Python 3.7+.

```python
#Three creation styles
empty_dict = {}

patient = {"id": "P0042", "age": 34, "temperature": 38.7, "admitted": True}

#From a list of (key, value) tuples
gene_data = dict([("BRCA1", "chr17"), ("TP53", "chr17"), ("KRAS", "chr12")])

#from two parallel lists using zip() - a preview of what zip() does
gene_ids   = ["G001", "G002", "G003"]
gene_names = ["BRCA1", "TP53", "KRAS"]
id_to_gene = dict(zip(gene_ids, gene_names))

print(f"patient:    {patient}")
print(f"gene_data:  {gene_data}")
print(f"id_to_gene: {id_to_gene}")
print(f"\nType: {type(patient).__name__}")
print(f"Keys:   {list(patient.keys())}")        #returns keys as list
print(f"Values: {list(patient.values())}")      #returns values as list
print(f"Pairs:  {list(patient.items())}")       #returns both keys and values as list
```

### 2. Accessing values

`Two ways` - and the difference matters. `d[key]` raises `KeyError` if the key doesn't exist. `d.get(key)` returns `None` (or a default you supply). In analytics pipelines, you almost always want `.get()`.

```python
patient = {"id": "P0042", "age": 34, "temperature": 38.7, "bp": 120}

# Direct access - fast, but fails on missing keys
print(f"Age: {patient['age']}")
print(f"Temp: {patient['temperature']}")

# .get() - safe access
print(f"\nBlood glucose: {patient.get('glucose')}")             # None
print(f"Blood glucose: {patient.get('glucose', 'not recorded')}")  # default

#chained .get() for nested dicts (safe multi-level access)
records = {"P0042": {"labs": {"HbA1c": 6.1}}}
hba1c = records.get("P0042", {}).get("labs", {}).get("HbA1c", "N/A")        #it returns an empty dictionary {} instead of None, .get('labs', {}): Looks for the labs key inside whatever was returned from the first step. If P0042 didn't exist, it now looks for labs inside the empty dictionary {}. If it's missing again, it returns another empty {}.
print(f"\nHbA1c for P0042: {hba1c}")

#key membership - use 'in', not .get()
print(f"\n'bp' in patient:      {'bp' in patient}")
print(f"'glucose' in patient: {'glucose' in patient}")
```

#### Reading all keys, values, and pairs

```python
#reading all keys, values, and pairs - using list comprehensions
lab_results = {
    "WBC": 8.2, "RBC": 4.9, "Hb": 13.5, "Plt": 250,
    "Na": 142,  "K": 4.1,   "Cr": 0.9,  "Glucose": 6.2
}

#build display strings using list comprehensions on .items()
lines = [f"  {k:<12} {v:.1f}" for k, v in lab_results.items()]
# print(lines)
print("Lab Results:")
print("\n".join(lines))

#check all vs any conditions using comprehensions
above_ref = [k for k, v in lab_results.items() if v > 10]
print(f"\nValues above 10: {above_ref}")
```

### **3. Mutating dictionaries**

Dictionaries are mutable - you can `add`, `change`, and `remove` entries after creation.

#### 3a. Adding or Updating Entries/Values

```python
patient = {"id": "P0042", "age": 34}

#add new key
patient["temperature"] = 38.7
print(f"Added temp:  {patient}")

#update existing key
patient["age"] = 35
print(f"Updated age: {patient}")

# .update() - merge another dict in (or keyword arguments)
vitals = {"bp_sys": 145, "bp_dia": 92, "hr": 88}
patient.update(vitals)
print(f"\nAfter update: {patient}")

# .setdefault() - add key only if it doesn't already exist
patient.setdefault("status", "admitted")   # sets it
patient.setdefault("age", 99)              # does NOT overwrite, does absolutely nothing to the dictionary when the key is already there
print(f"\nAfter setdefault: age={patient['age']}, status={patient['status']}")

print(f'\nComplete Dictionary:\n{patient}')
```

#### 3b. Removing Elements (Key and Val both)

```python
patient = {"id": "P0042", "age": 34, "temp": 38.7, "status": "admitted"}

# .pop(key) - remove and return a value
temp = patient.pop("temp")
print(f"Popped temp:     {temp}")
print(f"Patient now:     {patient}")

# .pop(key, default) - safe version (won't raise KeyError)
glucose = patient.pop("glucose", "not recorded")
print(f"\nGlucose (missing): {glucose}")

# .popitem() - removes and returns the LAST inserted item (Python 3.7+)
last_key, last_val = patient.popitem()
print(f"\nPopped last item: {last_key} = {last_val}")
print(f"Patient now: {patient}")

# .clear() - empty the dict entirely
temp_dict = {"a": 1, "b": 2}
temp_dict.clear()
print(f"\nAfter clear: {temp_dict}")
```

### 4. Nested dictionaries

Real-world data is nested. Patient records contain multiple sections. Gene databases contain annotations, coordinates, and expression data. Nested dicts model this naturally.

#### 4a. Accessing Elements from Nested Dictionary

```python
#hospital patient database
hospital = {
    "P001": {
        "name": "Ahmad Raza",
        "vitals": {"temp": 38.9, "bp_sys": 145, "hr": 92},
        "labs":   {"WBC": 12.1, "glucose": 9.4, "HbA1c": 7.2}
    },
    "P002": {
        "name": "Sara Khan",
        "vitals": {"temp": 36.7, "bp_sys": 118, "hr": 72},
        "labs":   {"WBC": 7.8, "glucose": 5.1, "HbA1c": 5.4}
    },
    "P003": {
        "name": "Bilal Ahmed",
        "vitals": {"temp": 39.4, "bp_sys": 158, "hr": 105},
        "labs":   {"WBC": 15.3, "glucose": 11.7, "HbA1c": 8.9}
    }
}

#access patterns - chain keys
print(f"P001 temperature:   {hospital['P001']['vitals']['temp']}")
print(f"P002 glucose:       {hospital['P002']['labs']['glucose']}")

# Safe nested access - chained .get() never raises KeyError
for_p999 = hospital.get("P999", {}).get("vitals", {}).get("temp", "N/A")
print(f"P999 temperature:   {for_p999}")
```

#### 4b. Extracting structured data using comprehensions

```python
hospital = {
    "P001": {"name":"Ahmad Raza","vitals":{"temp":38.9,"bp_sys":145,"hr":92},
             "labs":{"WBC":12.1,"glucose":9.4,"HbA1c":7.2}},
    "P002": {"name":"Sara Khan","vitals":{"temp":36.7,"bp_sys":118,"hr":72},
             "labs":{"WBC":7.8,"glucose":5.1,"HbA1c":5.4}},
    "P003": {"name":"Bilal Ahmed","vitals":{"temp":39.4,"bp_sys":158,"hr":105},
             "labs":{"WBC":15.3,"glucose":11.7,"HbA1c":8.9}}
}

#extract structured data using comprehensions
names_and_temps = [
    (pid, data["name"], data["vitals"]["temp"])
    for pid, data in hospital.items()
]

print(f'\nNames and Temps: {names_and_temps}')
print("Patient temperatures:")
print("\n".join(f"  {pid}  {name:<15}  {temp}°C" for pid, name, temp in names_and_temps))

#elevated glucose - comprehension filter
flagged = [
    (pid, data["name"], data["labs"]["glucose"])
    for pid, data in hospital.items()
    if data["labs"]["glucose"] > 7.0
]
print(f"\nFlagged (glucose > 7.0):")
print("\n".join(f"  {pid}  {name:<15}  glucose={g}" for pid, name, g in flagged))
```

### 5. Dictionary comprehensions

You learned list comprehensions in Class 3: `[expr for x in iter if cond]`.
Dict comprehensions are the same idea, just wrapping key-value pairs:
`{key_expr: val_expr for x in iter if cond}`

```python
#base case - transform a list into a dict
gene_names = ["BRCA1", "TP53", "KRAS", "EGFR"]
name_to_len = {g: len(g) for g in gene_names}
print(f"Name ---> length: {name_to_len}")

#from a list of tuples (key-value pairs)
raw_data = [("temp", 38.7), ("bp", 145), ("hr", 92), ("glucose", 9.4)]
vitals = {k: v for k, v in raw_data}
print(f"\nVitals: {vitals}")

#with a filter - only include elevated vitals
thresholds = {"temp": 37.5, "bp": 140, "hr": 100, "glucose": 7.0}
elevated = {k: v for k, v in vitals.items() if v > thresholds.get(k, float('inf'))}
print(f"Elevated markers: {elevated}")
```

#### 5a. Practical Use of dict comprehension

```python
#practical dict comprehension - frequency counting
text = "data analytics with python helps data-driven decisions in analytics"
words = text.replace("-", " ").split()

# {word: count} - use set() to deduplicate words, .count() to tally
freq = {word: words.count(word) for word in set(words)}
print("Word frequencies:")
# print(freq)

# Sort by frequency (dict.get as key - works without lambda)
ranked = sorted(freq, key=freq.get, reverse=True)
print("\n".join(f"  {word:<15} {freq[word]}" for word in ranked[:8]))

# Invert a dict (value ---> key) - useful when you need reverse lookup
gene_to_chr = {"BRCA1": "chr17", "TP53": "chr17", "KRAS": "chr12", "MYC": "chr8"}
chr_to_genes = {}
for chrom in set(gene_to_chr.values()):
    chr_to_genes[chrom] = [g for g, c in gene_to_chr.items() if c == chrom]
print(f"\nChromosome ---> genes: {chr_to_genes}")
```

## 6. Sets

A set is an **unordered collection of unique, hashable items**. The uniqueness is enforced automatically - adding a duplicate is silently ignored. Sets are built for three tasks:
1. Fast membership testing (O(1) vs O(n) for lists)
2. Deduplication
3. Mathematical set operations

```python
# Creation
nucleotides = {"A", "T", "C", "G"}
empty_set   = set()                    # NOT {} - that creates an empty dict
from_list   = set([1, 2, 2, 3, 3, 3]) # deduplication
from_string = set("ATCGATCG")          # unique characters

print(f"nucleotides: {nucleotides}")
print(f"from_list:   {from_list}")    # {1, 2, 3}
print(f"from_string: {from_string}")  # {'A','T','C','G'}
print(f"empty_set type: {type(empty_set).__name__}")

# Key property: sets are UNORDERED - don't rely on display order
sample = {3, 1, 4, 1, 5, 9, 2, 6, 5}
print(f"\nSet of [3,1,4,1,5,9,2,6,5]: {sample}")   # no duplicates, any order
```

### Mutation methods and Fastness Check

- .add() - add one item
-  .update() - add multiple items/expanding set
-  .discard() - remove element (no error)
-  .remove() - remove element (KeyError if missing)

```python
oncogenes = {"KRAS", "EGFR", "MYC"}

oncogenes.add("PIK3CA")            # add one item
print(f"After add:    {oncogenes}")

oncogenes.update({"BRAF", "RET"})  # add multiple/expanding a set
print(f"After update: {oncogenes}")

oncogenes.discard("XYZ")           # remove - no error if missing
oncogenes.remove("RET")            # remove - KeyError if missing
print(f"After remove: {oncogenes}")

# Membership testing - this is where sets shine
big_list = list(range(1_000_000))
big_set  = set(big_list)

import time
target = 999_999

t0 = time.time()
_ = target in big_list
list_time = time.time() - t0

t0 = time.time()
_ = target in big_set
set_time = time.time() - t0

print(f"\nMembership test on 1M items:")
print(f"  List: {list_time:.6f}s")
print(f"  Set:  {set_time:.6f}s")
print(f"  Set is {list_time/(set_time+1e-10):.0f}x faster")
```

# Speed Performance: Lists vs. Sets

<details>
<summary><b>1. The Grocery List vs. The Organized Library</b> (Click to expand)</summary>
<br>

*   **A List** is like a grocery list. To find if "milk" is on it, our eyes have to start at the top and scan down line by line. If we are looking for the very last item (999_999), we have to look through all one million items first.
*   **A Set** is like an organized library. Every book has a specific code based on its name. The computer uses a math formula to instantly know the exact shelf and row where that item belongs. It goes straight there.

</details>

<details>
<summary><b>2. How It Works Under the Hood</b> (Click to expand)</summary>
<br>

### Lists Use Linear Search
When we ask Python `target in big_list`, it starts at index 0 and checks every number in order:
*   Is it item #1? No.
*   Is it item #2? No.
*   ...
*   Is it item #1,000,000? Yes!

This takes a long time because the computer has to make one million checks.

### Sets Use Hash Tables
When we put a number into a set, Python turns that number into a "hash value" (a special address index).
*   When we check `target in big_set`, Python doesn't look at the other 999,999 numbers.
*   It computes the address for 999_999 instantly.
*   It jumps directly to that address.
*   It sees the number is there and says "True!"

Because it jumps straight to the answer, a set takes the same amount of time whether it holds 10 numbers or 10 billion numbers!

</details>

<details>
<summary><b>3. Understanding Hash Tables & Functions</b> (Click to expand)</summary>
<br>

The easiest way to understand a hash is to think of a library catalog system or a student ID card. 
A hash takes a piece of data (like a word or a number) and turns it into a unique index number that points to a specific location in memory.

### Example: The School Locker
Imagine a school with 1,000 students and 1,000 lockers.

*   **The List Way:** Alex walks down the hallway, reading the name tag on every single locker starting from locker #1 until he finds his name.
*   **The Hash Way:** The school uses a "hashing formula" based on the alphabet. The formula is: *Take the first letter of the last name, find its position in the alphabet, and multiply by 10.*
*   **The Execution:** "Smith" starts with S (the 19th letter). 
    $$19 \times 10 = 190$$
    Alex instantly knows his locker is #190. He walks straight to it.

In this example, the math formula is the **hash function**, and 190 is the **hash value**.

</details>

<details>
<summary><b>4. How the Set Uses This Hash</b> (Click to expand)</summary>
<br>

When you run `target in big_set`, Python executes these steps in the background:

1. It calculates `hash(999_999)`, which equals 999999.
2. It goes directly to the memory address container #999999.
3. It sees the number is sitting inside that container.
4. It instantly shouts "True!" without looking at anything else.

</details>

```python
# Hashing a number
print(hash(999_999))  
# Output: 999999 (Python is clever; small integers hash to themselves!)

# Hashing a text string
print(hash("apple"))  
# Output:  (A unique, giant index number)
```

### 7. Set operations

Four fundamental operations - all mirror mathematical set theory. Essential in comparative genomics, patient cohort analysis, and feature engineering.

- union - |
- intersection - &
- difference - -
- symetric difference - ^
- .issuperset()
- .issubset()
- .isdisjoint()

```python
# Two gene sets from different studies
study_A = {"BRCA1", "TP53", "KRAS", "EGFR", "MYC", "PIK3CA"}
study_B = {"BRCA1", "PIK3CA", "KRAS", "PTEN", "AKT1", "TP53"}

print(f"Study A: {study_A}")
print(f"Study B: {study_B}")
print()

# Union - all genes from either study
print(f"Union (|):       {study_A | study_B}")

# Intersection - genes in BOTH studies
print(f"Intersection (&):{study_A & study_B}")

# Difference - genes only in A (not in B)
print(f"A only (-):      {study_A - study_B}")
print(f"B only (-):      {study_B - study_A}")

# Symmetric difference - genes in one study but not both
print(f"Either not both: {study_A ^ study_B}")
```

```python
# Subset, superset, disjoint
core_pathway = {"BRCA1", "TP53"}
study_A = {"BRCA1", "TP53", "KRAS", "EGFR"}
unrelated = {"ACTB", "GAPDH"}

print(f"core_pathway ⊂ study_A: {core_pathway.issubset(study_A)}")    # True
print(f"study_A ⊃ core_pathway: {study_A.issuperset(core_pathway)}")  # True
print(f"study_A ∩ unrelated = ∅:{study_A.isdisjoint(unrelated)}")     # True (no overlap)

# Alternative syntax using operators
print(f"\nUsing operators:")
print(f"  core_pathway <= study_A: {core_pathway <= study_A}")  # subset
print(f"  study_A >= core_pathway: {study_A >= core_pathway}")  # superset
print(f"  core_pathway < study_A:  {core_pathway < study_A}")   # proper subset
```

#### Practical: Venn summary using set operations

```python
group_A = {"BRCA1","TP53","KRAS","EGFR","MYC"}
group_B = {"BRCA1","PIK3CA","KRAS","PTEN","TP53"}
group_C = {"KRAS","MYC","CDK4","RB1","TP53"}

core     = group_A & group_B & group_C          # in all three
only_A   = group_A - group_B - group_C
only_B   = group_B - group_A - group_C
only_C   = group_C - group_A - group_B
ab_only  = (group_A & group_B) - group_C
ac_only  = (group_A & group_C) - group_B
bc_only  = (group_B & group_C) - group_A

venn = {
    "Core (all 3)":   core,
    "A only":         only_A,
    "B only":         only_B,
    "C only":         only_C,
    "A & B only":     ab_only,
    "A & C only":     ac_only,
    "B & C only":     bc_only,
}

print("Venn summary:")
print("\n".join(f"  {k:<16} ({len(v):2d} genes): {v}" for k, v in venn.items()))
```

## 8. When to use dict vs list vs set

The right structure makes the code faster and clearer. Wrong choice = ugly workarounds.

```python
# Decision guide with timing

import time

# consider you have 50,000 patient IDs and need to check membership
patient_ids = [f"P{i:05d}" for i in range(50_000)]
patient_set  = set(patient_ids)
target = "P49999"

# print(patient_ids)

t0 = time.time()
result_list = target in patient_ids
list_ms = (time.time() - t0) * 1000

t0 = time.time()
result_set = target in patient_set
set_ms = (time.time() - t0) * 1000

print(f"50,000 patient lookup:")
print(f"  List membership: {list_ms:.4f} ms")
print(f"  Set membership:  {set_ms:.4f} ms")
print(f"  Same answer:     {result_list == result_set}")

# Summary table (printed, not shown as actual table structure)
summary = [
    ("Structure", "Order", "Duplicates", "Key lookup", "Best for"),
    ("list",      "Yes",   "Yes",        "O(n)",       "Ordered sequences, positional access"),
    ("dict",      "Yes*",  "Keys: No",   "O(1)",       "Key-value lookup, labelled data"),
    ("set",       "No",    "No",         "O(1)",       "Membership tests, deduplication"),
]
print()
for row in summary:
    print("  " + "  ".join(f"{cell:<20}" for cell in row))
```

### Practical 1 - The KeyError trap in a patient lookup system
**Context:** A clinic's dashboard looks up patient records by ID. What happens when a patient ID doesn't exist in the database?

```python
# Simulating a patient lookup - two approaches
records = {
    "P001": {"name": "Ahmad", "temp": 38.9},
    "P002": {"name": "Sara",  "temp": 36.7},
}

# Students predict: what does records["P999"] do?
print("Approach 1: Direct key access")
print("  records['P001']:", records["P001"])
print("  records['P999'] ---> ", end="")
try:
    print(records["P999"])
except KeyError as e:
    print(f"KeyError: {e}  <--- crashes the dashboard!")

# The fix - .get() with a default
print("\nApproach 2: .get() - safe lookup")
result = records.get("P999", {"name": "Unknown", "temp": None})
print(f"  records.get('P999', default): {result}")

# Real-world pattern: .get() with empty dict default for nested access
temp = records.get("P999", {}).get("temp", "N/A")
print(f"\n  Chained .get() for 'P999' temp: {temp}")
print("  Dashboard never crashes - returns 'N/A' instead")
```

### Practical 2 - Set deduplication speed vs list
**Context:** A genomics pipeline collects gene names from 10,000 sequencing reads. Many genes appear multiple times. How do you get the unique list?

```python
import time, random

# Simulate: 10,000 gene hits, drawn from a pool of 200 unique genes
gene_pool = [f"GENE{i:03d}" for i in range(200)]
raw_hits  = [random.choice(gene_pool) for _ in range(10_000)]

print(f"Total hits    : {len(raw_hits)}")
print(f"Expected unique: ~200")

# Method 1: manual deduplication with list
t0 = time.time()
unique_list = []
seen = []
for g in raw_hits:
    if g not in seen:      # O(n) per check
        seen.append(g)
        unique_list.append(g)
t1 = time.time() - t0

# Method 2: set - O(1) per insertion
t0 = time.time()
unique_set = list(set(raw_hits))
t2 = time.time() - t0

print(f"\nUnique genes found: {len(unique_set)}")
print(f"Manual loop time:   {t1*1000:.3f} ms")
print(f"set() time:         {t2*1000:.3f} ms")
print(f"set() is ~{int(t1/(t2+1e-9))}x faster")
print("\nLesson: set() is the correct tool for deduplication - always")
```

### Practical 3 - Dict comprehension vs manual approach
**Context:** You receive a CSV-parsed list of (gene, chromosome) pairs and need to build a lookup dict. Show the manual approach first, then the comprehension.

```python
# Raw data from a CSV parser (Class 2-style parsing)
raw_lines = [
    "BRCA1,chr17,43044295",
    "TP53,chr17,7661779",
    "KRAS,chr12,25398284",
    "EGFR,chr7,55019032",
    "MYC,chr8,127735434",
]

# Building the dict manually (conceptually - without for loop, using comprehension)
parsed = [line.split(",") for line in raw_lines]
print("Parsed rows:", parsed)

# Dict comprehension - reads almost like English
gene_chr  = {row[0]: row[1]     for row in parsed}
gene_pos  = {row[0]: int(row[2]) for row in parsed}
gene_info = {row[0]: {"chr": row[1], "pos": int(row[2])} for row in parsed}

print(f"\ngene_chr:  {gene_chr}")
print(f"gene_pos:  {gene_pos}")
print(f"\ngene_info['BRCA1']: {gene_info['BRCA1']}")

# Filtered: only genes on chr17
chr17_genes = {g: info for g, info in gene_info.items() if info["chr"] == "chr17"}
print(f"\nGenes on chr17: {list(chr17_genes.keys())}")
```

### Practical 4 - Set operations replace complex list logic
**Context:** Two departments each submit a list of drug targets. Management wants to know which are shared, which are unique to each team, and whether the teams are working in completely different areas.

```python
# Drug target lists from two research teams
team_alpha = {"EGFR", "KRAS", "BRAF", "MEK1", "ERK2", "PIK3CA"}
team_beta  = {"TP53", "RB1", "KRAS", "PIK3CA", "PTEN", "AKT1"}

# Questions management asks:
print("=== Drug Target Overlap Analysis ===")
print(f"Team Alpha targets: {len(team_alpha)}")
print(f"Team Beta  targets: {len(team_beta)}")

# Q1: Are teams working on completely different targets? (expensive with lists)
print(f"\nNo overlap at all? {team_alpha.isdisjoint(team_beta)}")  # False

# Q2: Shared targets - could combine resources
shared = team_alpha & team_beta
print(f"Shared targets ({len(shared)}): {shared}")

# Q3: Unique to each team - their competitive advantage
print(f"Alpha only ({len(team_alpha - team_beta)}): {team_alpha - team_beta}")
print(f"Beta only  ({len(team_beta - team_alpha)}): {team_beta - team_alpha}")

# Q4: Combined portfolio
total_coverage = len(team_alpha | team_beta)
print(f"\nCombined unique targets: {total_coverage}")
print(f"Overlap efficiency: {len(shared)/total_coverage*100:.0f}% redundancy")

# With lists, Q2 alone would take 10+ lines. With sets: 1 line.
```

## 9. Practical Exercise - Word Frequency Analyzer
**Difficulty: Medium**

Build a frequency analysis pipeline using only dict comprehensions and set operations. No explicit for loops or if statements yet - everything via comprehensions.

```python
'''In Python, when you place literal strings next to each other inside parentheses without commas, 
Python automatically merges them into a single continuous string at compile time. 
This is called implicit string literal concatenation.
In Python, commas make a tuple, not parentheses. 
If we added a comma between those strings, then it would become a tuple of strings'''
```

```python
# Dataset: biomedical abstract text
abstract = (
    "machine learning methods in genomics have enabled discovery of novel "
    "biomarkers machine learning algorithms trained on genomic data can predict "
    "drug response patterns in cancer genomics deep learning models for genomics "
    "have improved variant calling accuracy in clinical genomics pipelines "
    "machine learning and genomics together advance precision medicine research"
)

print(abstract)
print(type(abstract))
# Step 1: Clean and tokenise using string methods (Class 2 skills)
words = abstract.replace(",", "").replace(".", "").lower().split()
print(f"Total words: {len(words)}")
print(f"Unique words: {len(set(words))}")
```

```python
# Step 2: Frequency dict using dict comprehension + set + .count()
stop_words = {"and","in","of","to","the","a","an","for","on","with",
              "have","can","are","is","has","been","their"}

freq = {
    word: words.count(word)
    for word in set(words)
    if word not in stop_words
}

# Step 3: Sort by frequency (dict.get as key - no lambda needed)
ranked = sorted(freq, key=freq.get, reverse=True)

print("Top 15 words by frequency:")
print("-" * 35)
print("\n".join(
    f"  {i+1:2d}. {word:<20} {freq[word]:3d}  {'█' * freq[word]}"       #did you like that █? you can search it via google and copy paste "Full Block"
    for i, word in enumerate(ranked[:15])
))
```

```python
# Step 4: Set operations on vocabulary
# Compare vocabulary between two 'documents' (split the abstract in half)
half = len(words) // 2
doc1_words = set(words[:half])
doc2_words = set(words[half:])

# Remove stop words from both
stop = {"and","in","of","to","the","a","an","for","on","with","have","can"}
doc1_vocab = doc1_words - stop
doc2_vocab = doc2_words - stop

print(f"Doc 1 vocabulary: {len(doc1_vocab)} unique words")
print(f"Doc 2 vocabulary: {len(doc2_vocab)} unique words")
print(f"Shared core terms ({len(doc1_vocab & doc2_vocab)}): {doc1_vocab & doc2_vocab}")
print(f"Unique to Doc 1:  {doc1_vocab - doc2_vocab}")
print(f"Unique to Doc 2:  {doc2_vocab - doc1_vocab}")
```

```python
# Step 5: Inverted index - which positions does each word appear at?
# {word: [position1, position2, ...]} using dict comprehension
inverted_index = {
    word: [i for i, w in enumerate(words) if w == word]
    for word in set(words)
    if word not in stop_words and words.count(word) >= 2
}

print("Inverted index (words appearing 2+ times, with positions):")
# Sort by word for clean display
ranked_index = sorted(inverted_index.items(), key=lambda x: -len(x[1]))
print("\n".join(
    f"  {word:<20} positions: {positions[:8]}{'...' if len(positions)>8 else ''}"
    for word, positions in ranked_index[:8]
))
```

## Summary

| Concept | Key point |
|---|---|
| `d[key]` vs `.get()` | `d[key]` raises `KeyError` on miss; `.get(k, default)` never crashes |
| `.items()` | Returns key-value pairs - use in comprehensions for dict iteration |
| Dict comprehension | `{k: v for k, v in iter}` extends list comprehension syntax to dicts |
| `set()` for dedup | O(1) insertion means instant deduplication, even on millions of items |
| Set operations | `&` intersection, `|` union, `-` difference, `^` symmetric difference |
| `.isdisjoint()` | True if two sets share no elements - O(min(len(A),len(B))) |
| When to use each | List → order matters; Dict --> named lookup; Set → uniqueness/membership |
